# 10장 대외거래

## 1. 개요

- 최근 수출입 동향을 빠르게 파악하기 위해서 통관기준 수출입 자료를 이용하고 있으며, 우리 경제의 실질구매력 변화는 교역조건을 통해 판단
    - 원자료 외에 계절조정 일평균 수출입액도 함께 활용하고 있으며, OECD선행지수 등을 참고하여 대외경제여건을 평가
- 국가 간에 발생하는 모든 경제적 거래는 경상수지와 자본 및 금융계정으로 구성
    - 특히, 경상수지는 상품수지, 서비스수지, 본원이전 소득수지로 나누어지는데, 주로 상품수지와 서비스스지를 중심으로 분석

## 2. 수출입

### 2.1. 통관기준 수출입

**(정의 및 지표작성)**
- 통관기준 수출입 자료는 세관을 통과하는 물품을 기준으로 집계되며, 국민계정상의 수출입 자료와 집계방식이 다소 상이
    - 통관기준 수출은 본선인도가격(FOB)을 기준으로 집계되고, 수입은 본선인도가격에 운임 및 보험료도 포함한 가격(CIF)을 기준으로 집계
    - 통관기준 수출입은 수출(입)신고서 수리일을 기준으로 집계되는 반면, 국민계정상 상품 수출입은 실제 소유권 변동이 일어난 시점을 기준으로 집계

**(해석)**
- 관세청의 통관기준 수출입은 익월 1일경에 집계되어 발표됨에 따라 최근의 수출입 동향을 보다 빠르게 파악할 수 있음
    - 국민계정상 수출입 관련 자료는 분기별로 제공되므로 그에 따른 시차로 인해 수출입 관련 동향 및 경기 판단을 위해 적기에 이용되기 어려운 측면이 존재
        - 따라서 수출입 동향은 주로 통관기준 수출입 자료를 바탕으로 분석
    - 한편, 산업 및 지역별 자료를 이용하여 수출입동향에 대한 보다 미시적인 분석도 가능

### 2.2. 상품별 수출

**(정의)**
- 상품별 수출 자료는 MTI(Ministry of Trade and Industry)코드를 기준으로 분류 
    - MTI 코드는 산업통상자원부의 수출입 통계 분류기준이며, 산업별 수출입 상황을 파악하기 위해 품목코드인 HS코드를 재분류 한것
        - HS(Harmonized System)코드는 국제적으로 통용되는 코드로서, 국가별 실정에 맞추어 코드를 보다 세분화시키는데, 우리나라의 경우 10자리까지 확대시킨 HSK코드가 있음.
        - 이 외에도 UN에서 많이 사용하는 SITC, BEC 등이 있음
    - 통산적으로 MTI 2~3단위를 기준으로 비중이 큰 상위 10여개 품목을 주로 관찰
        - 해마다 상위 품목들의 비중을 관찰하되 품목별 비중의 변화가 장기적인 경우 수출입 동향의 주요 지표에 포함된 품목을 변경

**(지표작성)**
- 일평균 수출액은 총수출액을 해당 월의 일요일과 공휴일을 제외한 조업일수로 나누어 계산
    - 산업통상자원부는 토요일을 0.5일로 간주하므로 일평균 수출액을 분석할때 주의
    - 한편, 계절조정 자료는 E-views 프로그램의 X-12모형을 이용하여 자체적으로 계산할 수 있음.

**(해석)**

**(출처 및 공표일시)**
- 매월 1일경 산업통상자원부에서 전월의 수출 잠정치를 발표하고 있으며, 20여일 후에 무역협회, IT통계포탈, 한국석유공사 등을 통해 확정치를 확인
    - 산업통상자원부는 보도자료의 형태로 전월 말일까지의 총수출액과 품목별 수출액의 잠정치를 제공

### 2.3. OECD 선행지수(OECD Composite Leading Indicator)

**(정의 및 지표작성)**
- 세계경제 및 주요국의 전반적인 실물경기 흐름을 예측하는 선행지표로서, 실제 경기순환의 저점과 정점을 6~9개월 선행하는 것이 목표
    - OECD 선행지수는 경기순환을 선행할 수 있도록 이와 유사한 주기적인 움직임을 보이는 경제 시계열 변수의 움직임을 종합하여 작성
        - 이를 위해 기준지표(reference series)로는 2012년 4월부터 GDP를 사용
    - Amplitude adjusted 지수는 장기 추세가 제거된 기준시계열(GDP)을 선행하는 지표
        - 장기추세 100을 기준으로 지수가 100이상이면 경기확장 국면, 100이하이면 경기하강 국면으로 평가
    - Trend Restored 지수는 본래의 기준시계열인 GDP와 비교 가능한 것으로서 Amplitude adjusted 지수에 추세를 포함
        - 기준시계열이 GDP로 변경된 이후 Trend restored 지수를 집계하는 과정에서 시차가 확대되었으며, 자료의 유효성을 확보하기 위해 Trend restored지수를 Amplitude adjusted 지수로 대체하여 활용

**(해석)**
- 물량기준 상품 수출은 궁극적으로  해외수요에 의해 결정되며, 상품 수출의 최근 추이 및 단기 전망을 위한 선행지표로서 OECD경기선행지수를 이용
    - OECD회원국이 세계경제에서 차지하는 비중을 고려할 때, OECD경기선생지수는 세계경제상황을 파악하기 위한 핵심 선행지표로 활용 될 수 있음.
        - 또한 OECD 경기선행지수는 월별자료로 제공되므로, 국민계정상 상품수출에 대한 선행지표로도 활용이 가능
    - OECD 경기선행지수 외에도 주요 신흥시장 6개국(중국, 브라질, 인도, 러시아, 인도네시아, 남아프리카공화국)을 포함한 OECD+NME(Non-OECD Member Economies)경기 선행지수와 NME 경기선행지수도 함께 관찰
    - 한편, 수년 전에는 선행지수의 전년동월대비 증가율이 총수출에 약 1개월 선행하였으나, 최근에는 거의 동행하는 모습을 보이고 있음. 

**(출처 및 공표일시)**
- OECD 경기선행지수는 매월 10일 전후로 발표되며, 수출자료와 2개월 정도의 시차가 존재
    - 특히, Trend restored 지수는 각국의 GDP가 발표된 이후에야 집계가 이루어지면서 Amplitude adjusted 지수보다 시차가 더 큼
        - OECD 선행지수 경로: https://stats.oecd.org/Index.aspx?DataSetCode ---- Monthly Economic Indicators ---- Composite Leading Indicators

**(조사주기 및 자료의 형태)**

### 2.4. 세계교역량지수

**(정의 및 지표작성)**
- 세계 교역량지수는 전체 교역의 99%를 차지하는 81개 국가들의 무역량을 기초로 작성되기 때문에 국제무역에 대한 대표성이 높은 지표이며 대외거래의 월별 동향을 파악하는 것이 목표
    - 세계교역량지수는 물량(volumes)과 가격(prices)을 기준으로 작성되며 기준연도(2010=100) 대비 국제무역량의 증감을 나타냄

**(해석)**
- 세계교역지수를 통해 물량기준 수출 동향 추이를 파악하고 대외 경제상황 전반을 점검
    - 수출경기를 예측하기 위한 보조지표로 OECD 경기선행지수와 함께 활용
    - 예전에는 세계교역량지수의 증가율이 총수출과 유사한 추이를 보였으나 근래에는 수출물량지수와 비슷한 움직임을 보이는 추세

**(출처 및 공표일시)**
- 매월 20일 경 네덜란드 경제정책분석국(CPB)에서 발표되며, 수출 자료와 2개월 정도의 시차가 존재
    - 전 기간의 자료가 소급 변경되기 떄문에 매월 전체 시계열 자료를 다시 확인할 필요
        - 세계교역량지수 경로: https://www.cpb.nl/en/ ---- World trade monitor

**(조사주기 및 자료의 형태)**
- 세계교역량지수는 2000년 1월이후의 월별 시계열 데이터를 제공
    - 물량(volume) 지준 지수는 계절조정된 자료이며, 추가적으로 일부 국가들의 고유한 특수성을 반영
    - 한편, 물량(volumes) 및 가격(prices) 기준으로 작성된 지수 외에도 대상국가들의 수출 및 수입을 지수화한 자료도 함께 제공

### 2.5. 용도별 수입

**(정의)**
- 수입은 MTI코드 기준의 품목별 분류 외에, 소비재와 원자재, 자본재로 구분되는 용도별 분류를 포함
    - 이에 따라 수입동향 지표에서는 용도별 수입과 원유 수입, 원유의 도입물량 및 도입단가 등을 함께 표기

**(지표작성)**
- 일평균 수입액은 앞서 설명한 일평균 수출액과 동일한 방법으로 계산
    - 계절조정 일평균 수입액도 E-views의 X-12 모형을 이용
    - 분기말이나 연말에 해당되는 월별자료가 발표되는 시점에는 산업통상자원부 보도자료에 용도별 수입과 지역별 수출입의 20일 기준 잠정치만 발표되므로 해당분기와 연도의 증가율을 계산하기 위해서는 전년동월의 20일 기준 수치가 필요
    - $$2022년 3/4분기 원자재 수입증가율 = (\frac{2022년 3/4분기 원자재수입액}{2021년 3/4분기 원자재수입액}-1)\times 100$$
        - $$2022년 3/4분기 원자재수입액 =  2022년7월~9월20일까지의 수입액$$
        - $$2021년 3/4분기 원자재수입액 =  2021년7월~9월20일까지의 수입액$$

**(해석)**
- 일부 통관기준 수입 자료는 상기한 속보성을 바탕으로 국내 주요 실물변수의 선행지표로 활용되기도 함.
    - 기계류 수입액이나 내수용 자본재 수입액 등은 국내 설비투자에 대한 유용한 선행지표로 평가되고 있음
    - 수입액 중 원유, 가스, 석유제품 등 주요 에너지 자원 수입이 미치는 영향을 관찰하기 위하여, 주요 에너지원 수입액과 이들을 제외한 총수입액 추이를 함께 확인

## 3. 무역지수 및 교역조건

### 3.1. 무역지수

**(정의)**
- 무역지수는 수출입 금액의 변동요인을 가격요인과 물량요인으로 구별하여 검토하기 위해 작성되는 지수로 수출입물량지수, 수출입금액지수, 수출입물가지수가 있음.
    - 관세청 통관수출입 통계를 이용하여 작성되고 있으며, 특수한 거래 형태를 보이는 선박, 무기류, 항공기, 예술품 등은 모집단에서 제외
    - 수출은 FOB가격, 수입은 CIF가격을 기준으로 조사되고 있으나, 계약통화와 원화를 기준으로한 지수도 함께 작성되고 있음.

**(지표작성)**
- 수출입물량지수는 수출입금액지수를 수출입물가지수로 나눈 값으로 정의
    - 한국은행에서 공표되는 수출입물가지수는 계약시점가격기준으로 작성되므로, 물량지수를 계산할 때에는 통관시점을 기준으로 하는 수출입금액지수에 맞추기 위해 시차가 조정된 수출입물가지수를 이용
        - $$수출(입)물량지수 = \frac{수출(입)금액지수}{수출(입)물가지수(시차적용)}\times 100$$

### 3.2. 교역조건 
**(정의)**
- 순상품교역조건지수는 수출물가지수와 수입물가지수의 비율로 정의
    - 여기서 수출입물가지수는 통관시점으로 시차조정이 된 지수를 이용
        - $$순상품교역조건지수 = \frac{수출물가지수(시차적용)}{수입물가지수(시차적용)}\times 100$$
    - 소득교역조건지수는 순상품교역조건지수에 수출물량지수를 곱한 후 100으로 나누어 계산
        - $$소득교역조건지수=\frac{수출금액지수}{수입물가지수(시차적용)}\times 100$$ 
        $$= \frac{수출물가지수(시차적용)\times 수출물량지수}{수입물가지수(시차적용)}$$
        $$ = \frac{순상품교역조건지수\times 수출물량지수}{100}$$

**(해석)**
- 순상품교역조건지수는 수출 1단위로 수입할 수 있는 상품의 양을 나타내며, 우리경제의 전반적인 대외구매력 변화를 파악할 수 있는 지표로 평가되고 있음.
    - 소득교역조건지수는 물량요인이 포함된 수출총액으로 수입할 수 있는 양을 나타냄.

    - 자본거래의 총액자료를 얻기 어렵지만 일반적으로 순액변동 분석만으로도 자본흐름에 대한 의미있은 정보를 얻을 수 있음.
    - 직접투자는 상대적으로 변동이 크지 않은 편이며, 파생금융상품은 환율이 급변하는 시기에는 상당한 변화가 발생할 수 있음.

### 5.2. 외환보유액

**(정의)**
- 중앙은행과 정부가 언제든지 사용할 수 있는 대외 외화금융자산이며, 높은 유동성과 시장성이 특징
    - 긴급 상황으로 금융기관 등 경제주체의 대외 결제가 어려워질 경우 또는 외환시장에서 환율이 급격하게 상승할 경우 시장안정을 위해 사용
    - 외환보유액은 정부(외국환평형기금) 및 한국은행이 보유하고 있는 유가증권(국채, 정부기관채, 금융채, 자산유동화증권 등), 예치금, SDR(특별인출권), IMF포지션(IMF 가맹국이 출자금 납입으로 보유하게 되는 IMF로부터의 교환성통화인출 권리), 화폐금(Monetary Gold) 등으로 구성
    - 한편, 외환보유액 증감에는 외화자산 운용 결과 뿐만 아니라 환율 변동에 따른 평가손익도 반영됨.

**(지표작성)**
- 한국은행은 1999년 8월까지는 총 외환보유액 기준으로 작성하고, 그 이후로는 가용 외환보유액 기준(총 외환보유액에서 국내은행의 해외점포 예치금 등을 제외한 금액)으로 작정

**(해석)**
- 국가의 대외 지불능력을 나타내는 지표로 외환보유액이 너무 적을 경우에는 대외채무를 상환하지 못하는 사실상의 국가부도 상태에 빠질 우려가 있음.
    - 충분한 외환보유액을 유지함으로서 국가신인도를 높여 기업 및 금융기관의 해외자본조달 비용을 낮추고 외국인 투자를 촉진시킬 수 있음.
    - 적정한 외환보유액 수준 유지에 대한 논란도 일부 제기되고 있으나, 소규모 개방경제와 지정학적 요인 등 우리경제의 특수성을 감안하년 충분한 수준의 외환보유액을 유지하는 것이 대외 신인도 유지와 위기재발 방지에 매우 중요
    - 한편, 외환보유액 규모가 커지면서 운용수익률에 대한 관심이 증가하는 추세

**(출처 및 공표일시)**
- 한국은행은 매월 5일경 전월 말의 외환보유액을 보도자료로 발표하고 있으며, 경제통계시스템에서도 관련 자료를 함께 제공
    - 외환보유액 경로: https://ecos.bok.or.kr/---- 통계검색 ---- 3.환율/통관수출입/외환보유액 ---- 3.5. 외환보유액

### 5.3. 국제투자대조표

**(정의)**
- 국제투자대조표는 특정시점에 거주자의 비거주자에 대한 금융자산(대외투자) 및 금융부채(외국인투자) 잔액과 그 변동내역을 체계적으로 기록
    - 상품형태 기준으로는 직접투자, 증권투자, 파생금융상품, 기타투자 및 준비자산 등으로 구성되며 각 세부 항목에 대한 자세한 내용은 국제수지 금융계정 부분을 참고
    - 대외채권 및 채무 통계 자료는 거래주체별, 기간별 등으로 분류 
        - 거래주체의 성격에 따라 중앙은행, 일반정부, 예금취급기관, 기타 등 4개 부문으로 분류
        - 금융상품의 만기를 기준으로 1년이하는 단기, 1년 초과는 장기로 구분

**(지표작성)**
- 대외채권 및 채무 통계자료는 기말 시점의 자산과 부채의 잔액을 시가로 평가하여 계산
    - 미 달러화 이외의 통화로 표시된 자산과 부채는 기말시점의 대미 환율을 이용함

**(해석)**
- 대외금융자산 및 부채의 종합적인 정보를 제공함으로써 우리나라의 외환건전성 등 대외지급능력을 판단하는 기초 자료로 활용
    - 경제규모가 확대됨에 따라 대외 채무와 채권은 모두 증가하는 것이 일반적인 현상
    - 따라서 외채의 수준과 대외지급능력을 정확히 파악하기 위해서는 총외채대비 단기외채 비율, 외환보유액 대비 단기외채 비율 또는 외환 보유액 대비 유동외채비율 등을 함께 검토하는 것이 바람직

**(출처 및 공표일시)**
- 한국은행에서 매 분기마다 국제투자대조표를 작성하여 2개월 후에 잠정치를 공표하고 있으며, 경제통계시스템에서 관련 자료를 모두 제공
    - 국제투자대조표 경로: https://ecos.bok.or.kr/---- 통계검색 ---- 2. 국민계정---- 2.6. 국제투자대조표/대외채권채무 ---- 2.6.1. 국제투자대조표(IIP)
