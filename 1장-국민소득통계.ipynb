# 1장 국민계정

## 1. 설명 

- **국민계정(국민소득통계)** 는 경제활동별 최종생산물이나 소비, 투자, 수출 등 최종 지출을 대상으로 국민경제 전체의 활동 수준과 흐름, 그리고 국민의 생활 수준을 보여 줌과 동시에 경제주체들의 의사결정과 사후평가 등에 가장 광범위하게 이용되는 대표적인 경제통계임.

- 국민계정통계 추계방법에는 생산접근법, 소득접근법, 지출접근법 등 크게 세 가지가 있는데 「국민소득 3면 등가의 원칙」 에 의거 어느 방법을 이용하여도 이론상으로는 동일한 결과를 얻을 수 있음.
    - 생산국민소득은 산출물을 생산하는데 투입한 중간재화와 서비스를 차감한 부가가치액이며, 생산접근법은 생산 및 공급 관련 통계가 비교적 잘 정비된 국가에서 산업연관표와 연계하여 널리 이용되는 방법임.
    - 분배국민소득은 가계 및 기업 등 각 경제주체들의 소득을 모두 합한 값이며, 소득접근법이라고 부르며 급여통계, 산업별 취업통계 등 소득수취측 기초자료를 이용하여 국민소득을 추계하는 방법으로 기초자료 사정상 이를 이용하는 나라는 많지 않음.
    - 지출국민소득은 소비, 투자(설비투자, 건설투자, 지식재산생산물투자), 수출입 등 각 경제주체들의 최종생산물에 대한 수요를 합한 값으로, 지출접근법은 소비, 투자, 재고증가 등 지출관련 자료를 기본으로 하여 지출측면에서 국민소득을 추계하는 방법임.
        - 우리나라는 생산 및 공급관련 통계가 상대적으로 잘 발달되어 있어 오래전부터 생산접근법으로 국민소득통계를 추계해 왔으며 2009년 기준년(2005년) 개편시 지출접근법을 도입하였음.
        
- 국민소득통계 작성시 첫 단계는 통계작성에 필요한 관련 기초통계자료의 수집임.
    - 국민소득통계는 국민경제 전체의 움직임을 체계적이고도 종합적으로 파악하는 가공통계이므로 170여종의 단위기초통계가 이용되고 있으며, 체계화되지 않은 단순자료를 포함하면 기초통계자료의 수는 300여종을 초과함.
        - 국민소득통계는 정합성을 갖춘 기초통계자료를 적시에 얼마만큼 많이 수집하여 이용하느냐에 따라 통계의 질이 좌우되기 때문에 기초통계자료의 개선, 개발과 함께 수집에 상당한 관심과 노력을 경주하고 있음.
        - 분야별로 다양한 형태의 기초통계자료가 수집되면 이의 오류를 확인한 다음 상품기준의 산출액, 중간투입액 및 부가가치를 추계함.
        - 각종 기초통계자료로부터 총산출(CT: Control Totals)을 먼저 확정한 다음 중간투입구조를 조사하여 중간투입액(IC: Intermediate Consumption)을 구하고, 총산출액에서 중간투입액을 차감하여 부가가치(VA: Value Added)를 산출
    - 추계 결과의 경제현실과의 부합 여부 등 적정성 검토를 거친 후 생산국민소득을 확정함.
        - 적정성 검토는 생산접근법 및 지출접근법에 의한 추계 결과가 전체 국민계정체계와의 내적 정합성과 광공업생산지수를 비롯한 기초통계와의 외적 정합성을 갖는 지를 종합적으로 검토
    - 한편 분배국민소득은 생산계정에서 산출된 부가가치를 토대로 사후적으로 소득의 발생과 사용을 각 제도부문(경제주체)에 따라 배분한 것으로 연간으로만 편제

## 2. 주요 개념

- **명목 국내총생산**(nominal Gross Domestic Product, GDP valued at current prices): 경제규모 등의 파악에 이용되는 지표로서 국내에서 생산된 최종생산물의 수량에 그때의 가격을 곱하여 산출하므로 명목GDP의 변동분은 최종생산물의 수량과 가격변동분이 혼재되어 있음
    - 국내라 함은 한 나라의 경제적 영역(economic territory)내를 의미하며, 이는 지리적 영역(geographical territory)에 기초하지만 반드시 일치하는 것은 아님

- **실질 국내총생산**(real Gross Domestic Product, GDP valued at chained xxxx year prices): 국내경제의 생산활동 동향을 나타내는 경제성장률 산정에 이용되는 지표로서 지수기준년부터 당해년까지 매년의 개별 재화와 서비스의 가중치 및 수량변화를 반영해 측정한 연쇄 물량지수에 지수기준년(xxxx년, 현재 2000년 이후 데이터는 2015년)의 GDP 금액을 곱하여 산출한 물량측정치임. 이에 따라 실질GDP의 변동분은 가격 변화분을 제거한 순수한 생산수량의 변동분만을 나타냄
    - 일반적으로 실질GDP의 추계는 연쇄가중법을 이용하여 측정
        - 연쇄가중법은 OECD 회원국 등 대부분의 선진국들이 채택하고 있는 방식으로 GDP 추계시 기준년이 매년 직전년도로 변경               
            - 연쇄가중법에서는 매년 직전년도를 기준년으로 삼아 당해년의 전년대비 물량증가율(연환산지수)을 먼저 구하고 이를 누적하여 당해년의 연쇄지수와 실질GDP를 사후적으로 계산
            - 최근의 경제상황과 관련 높은 직전년도의 가격 또는 가중치를 사용하여 각 년도의 정확한 경제성장률 측정에 중점
            - 고정가중법에 비해 경제현실의 반영도가 높고 국제비교가 용이하다는 장점이 있으나, 총량금액과 그 구성항목 합계간의 가법성이 성립하지 않는다는 단점이 존재
            - 따라서 하위 시계열의 합산을 통해 상위시계열을 계산하거나 상위시계열과 일부 하위시계열을 통해 그 외의 하위시계열을 계산하는 경우에는 한국은행에서 제공하는 연쇄시계열 합산 프로그램을 이용하여 계산 
            
- **실질 국내총소득**(real Gross Domestic Income): 국내에서 생산된 최종생산물의 실질 구매력을 나타내주는 지표로서 명목GDP로부터 개별상품의 절대가격 변화에 따른 변동분을 제거한 것이 실질GDP(물량측정치)가 되며, 실질GDP에서 교환되는 상품간의 상대가격 변화에 따른 구매력의 변동분(실질거래손익)을 조정하여 실질 GDI(구매력 측정치)를 구함

    - 실질 GDI = 실질 GDP + 교역조건 변화에 따른 실질무역손익

- **명목 국민총소득**(nominal Gross National Income): 1인당 국민소득, 국가경제규모등을 파악하는 데 이용되는 지표로 우리나라 국민이 국내는 물론 해외에서 생산활동에 참여한 대가로 벌어들인 명목 총소득을 의미하며, 명목GDP에 명목 국외순수취요소소득을 더하여 산출(종전의 명목 GNP)
    - 1인당 국민소득(GNI)은 국민들의 생활수준을 나타내는 지표로서 연간의 명목국민총소득을 추계인구(매년 7월 1일 기준)로 나누어 구하며 국제비교를 위해 미달러화(연평균환율 적용)로도 표시
    
- **국민총처분가능소득**(Gross National Disposable Income): 소비율, 투자율, 저축률등의 산정에 이용되는 지표로서 소비나 저축으로 자유로이 처분할 수 있는 총소득으로 GNI에 국외순수취경상이전을 더하여 산출하며 일반적으로 명목지표만 편제

## 2. 국내총생산(Gross Domestic Product: GDP)

- **국내총생산(Gross Domestic Product, GDP)** 은 한 나라의 영토 내에서 가계, 기업, 정부 등 모든 경제주체가 일정기간동안 생산한 재화와 서비스의 시장가치의 합으로 한 나라의 생활수준(living standard)을 나타내는 대표적인 거시지표임

### 1. 경제활동별 국내총생산
- **(정의)** 산출물에서 생산에 투입된 중간재화및 서비스를 차감한 부가가치액인 생산국민소득을 15개의 경제활동별(산업별)로 분류하여 집계
    - 생산국민소득은 `2008 SNA`의 계정체계 중 생산계정에서 도출되는 국민소득
- **(지표작성)** 
    - 산출액, 중간소비 및 부가가치는 기본적으로 국민소득 분류체계상의 상품(384부문)별로 추계한 후 대산업 수준에서 15개의 산업기준으로 전환하여 경제활동별 금액을 결정
        - 15개의 산업기준은 농림어업, 광업, 제조업, 전기⋅가스⋅수도사업, 건설업을 비롯하여 10개의 서비스업(도소매및음식숙박업, 운수및보관업, 금융및보험업, 부동산및임대업, 정보통신업, 사업서비스업, 공공행정및국방등정부서비스, 교육서비스업, 보건및사회복지서비스업, 문화및기타서비스업)으로 분류
        - 기초가격으로 평가된 산업별 부가가치의 합, 즉 총부가가치(GVA: gross value added)에 순생산물세(생산물세 - 생산물보조금)을 더해야 시장가격으로 평가되는 GDP와 일치함.
- **(해석)**
    - 산업별 성장률과 전반적인 경기를 판단하는 과정에서 경제활동별 GDP를 이용
        - 국내총생산 자료는 분기별로 발표되므로 적시성이 떨어지는 단점이 있으나, 매월 발표되는 생산지수를 통해 해당분기의 부문별 성장율을 예측하고 이를 단기 경제성장률 전망에 활용
            - 예를 들어, 4~5월의 생산지수를 이용하여 2/4분기 부문별 성장률을 예측하고 이를 합산하여 2/4분기 전체 성장률을 전망할 수 있음.
- **(출처 및 공표일시)**

        - 연간 국민계정 통계는 잠정치의 경우 해당년도 종료 후 3개월이내에, 확정치의 경우 익익년 3월에 공표

### 3. GDP 디플레이터

- **(정의 및 지표작성)**
    - GDP 디플레이터는 국내에서 생산된 모든 재화와 서비스의 물가수준을 측정하는 지표로, 명목GDP를 실질GDP로 나누어 사후적으로 얻어지는 값
    - 국민계정체계에서 GDP 디플레이터는 명목GDP와 실질GDP의 비율로 다음과 같이 계산
    $$GDP 디플레이터 = \frac{명목 GDP}{실질GDP}\times 100$$
        
        - 명목GDP는 당해년에 국내에서 생산된 최종 생산물의 수량에 당해년의 가격을 곱하여 산출하며, 국민경제의 구조나 국민소득을 구성하는 각 요소들의 중요성이 시간의 경과에 따라 어떻게 변동하였는가를 분석할 때 사용
        - 실질GDP는 당해년에 국내에서 생산된 최종 생산물의 수량에 기준년(현재 2015년)의 가격을 곱하여 산출하며, 주로 장기간에 걸친 물량 수준 및 투입구조의 변동을 분석하고 경제성장률을 측정하는데 사용
        
- **(해석)**
    - 물가수준 파악과 성장기여도 및 경기예측에 필요한 핵심지표로 국민소득에 영향을 주는 모든 물가요인을 포괄하는 종합적인 물가지수임.
        - 명목GDP의 변동에는 생산물의 수량과 가격변동분이 혼재되어 있는 반면, 실질GDP는 생산수량으로 나타낸 소득이므로 실질GDP의 변동은 물가변동에 의한 증감액이 제거된 생산물 수량의 변동만을 나타냄.
        - 따라서, 명목GDP와 실질GDP의 차이는 국민경제 전체의 물가수준으로 설명
        - GDP디플레이터는 경제활동별 성장기여도와 지출항목별 성장기여도를 산출에도 사용

- **(출처 및 공표일시)**
    - 한국은행에서는 잠정치 통계공표시 분기 경제활동별 GDP디플레이터와 지출항목별 GDP디플레이터를 발표

### 4. GDP에 대한 성장 기여도

- **(정의)**
    - 특정 경제활동 부문이나 지출항목이 전체 GDP 성장률에 어느정도 기여를 했는지 분석하는 지표
        - 따라서 각 부문별 경제활동의 성장기여도 합은 해당기간의 전체 GDP 성장률과 일치
- **(지표작성)**
    - 실질GDP 통계가 연쇄가중방식으로 추계됨에 따라 연간 성장기여도와 분기성장기여도의 계산방법이 상이함.
        - 자세한 내용은 *연쇄시계열의 성장기여도 계산방법* (국민계정리뷰 2008년 제4호)를 참조
    - 한편, 한국은행에서 제공하는 연쇄시계열 합산 프로그램과 성장기여도 계산프로그램을 이용하여 순수출, 내수, 내수(재고제외) 등의 GDP에 대한 성장기여도를 계산할 수 있음.
        - 순수출의 성장기여도는 수출 성장기여도에서 수입 성장기여도를 차감
        - 내수 성장기여도는 민간소비와 정부소비, 총자본형성의 성장기여도를 합산
        - 내수(재고제외)의 성장기여도는 내수의 성장기여도에서 재고증감의 성장기여도를 차감
- **(해석)**        
    - 성장기여도 지표는 특정 경제활동 부문이나 특정 지출항목이 전체 경제성장에 어느정도 기여했는가를 나타내는 지표로 해당기간의 경제성장을 어느 항목이 주도했는가를 파악하는데 유용
        - 성장기여도를 이용하여 우리 경제의 성장과정을 평가할 수 있으며, 향후 정부의 정책방향 등에도 반영될 수 있음.
- **(출처 및 공표일시)**        
    - 한국은행에서는 GDP 공표시 경제활동별 GDP 및 지출항목별 GDP에 대한 성장기여도 자료를 함께 발표

### 5. 기타 국민소득 관련 지표
#### 5.1 실질 국내총소득 (Gross Domestic Income: GDI)
- **(정의 및 지표작성)**    
    - 실질 GDP에 교역조건 변화에 따른 실질 무역손익을 합한 값을 의미
        - 교역조건 변화에 따른 실질 무역손익은 국가 간의 거래에서 교역조건이 변화함에 따라 발생하는 실질소득의 국외유출 및 국내유입을 의미
            - 교역조건이 변화하면 실질 무역이익 또는 무역손실이 발생하여 실질GDP와 실질소득수준간에 차이가 발생
        - 교역조건의 변화를 반영한 총 실질 무역손익에서 일정시점(기준년)으로부터 교역조건이 변화하지 않았다고 가정하였을 겨우의 실질 무역손익을 차감하여 계산
        $$ T = \frac{X-M}{P^{*}}-(x-m)$$
        $T$: 교역조건 변화에 따른 실질 무역손실<br>
        $X-M$: 명목 무역손실 [당해년 재화 및 서비스의 수출입차]<br>
        $P^{*}$: 환가지수 [수출입가격지수의 평균 $=(P_x+P_m)/2$]<br>
        $x-m$: 교역조건 불변 실질 무역손익 [기준년가격 재화와 서비스의 수출입차 $=X/P_x-M/P_m$]<br>
        $P_x$: 수출가격, $P_m$: 수입가격      
        
- **(해석)**    
    - 실질 GDI는 국내에서 생산된 최종생산물의 실질 구매력을 반영하는 소득지표
        - 수출품가격이 상승하면 국외에서 유입되는 소득이 늘어나면서 GDI는 증가하고, 수입품 가격이 상승하면 국외로 유출되는 소득이 늘어나면서 GDI는 감소
        - 교역조건의 변화에 따른 실질 무역손익이 반영된 실질GDI와 실질소비 및 투자 등 여타 부문별 지표를 단순하게 비교하여 경제현상을 파악하는 것은 다소 무리가 있음.
- **(출처 및 공표일시)**

#### 5.2 국민총소득 (Gross National Income: GNI)
- **(정의 및 지표작성)**    
    - 특정 경제적 영역내에서 주된 경제활동을 수행하는 거주자(주로 자국민)가 생산활동에 참여한 대가로 받은 소득의 합계
        - 실질구매력을 반영하는 소득지표로 명목 국민총소득과 실질 국민총소득으로 나눔.
            - 명목 국민총소득(GNI)은 명목GDP와 명목 국외순수취요소소득을 함한 값
            - 실질 국민총소득은 실질 국내총소득(GDI)과 실질 국외순추취요소소득을 합한 값
       - 국외순수취요소소득은 거주자가 해당 경제적 영역 외부(주로 국외)에서 생산요소를 제공한 대가로 받은 국외수취요소소득에서 비거주자가 생산활동에 참여하여 발생한 국외지급요소소득을 차감하여 계산
           - 거주성은 경제적 이익의 중심 및 생산설비 속지주의를 기준으로 판단
           
- **(해석)**    
    - 실질 국민총소득은 거주자가 국내는 물론 국외에서 벌어들인 소득의 실질 구매력을 나타내는 지표임.
        - 명목 국민총소득은 1인당 국민소득, 국가경제의 규모등을 파악하는 데 이용되는 지표로 거주자가 국내는 물론 해외에서 생산활동에 참여한 대가로 벌어들인 명목 총소득을 의미 (종전의 명목 GNP)
        - 1인당 국민소득(GNI)은 국민들의 생활수준을 나타내는 지표로서 연간 명목 국민총소득을 추계인구(매년 7월1일 기준)로 나누어 구하며, 국제비교를 위해 미 달러와(연평균환율 적용)로도 표시
        
- **(출처 및 공표일시)**    
    - 한국은행에서는 분기별 잠정치 통계발표시 국민총소득을 함께 발표

#### 5.3 국민순소득 (Net National Imcome: NNI)
- **(정의 및 지표작성)**    
    - 국민순소득(NNI)은 국민총소득에서 고정자본소모를 제외한 값
        - 고정자본소모는 생산활동에서 발생하는 생산에 이용한 기계 등 자산의 가치 감소분을 뜻함.
        - 일정기간 중에 생산된 생산물의 순수한 가치를 평가하기 위해 현재의 생산능력이 저하되지 않고 그대로 유지된 상태에서 생산이 이루어지는 것을 가정해야할 필요가 있으며, 이를 위해 국민총소득에서 고정자본소모만큼 조정을 하여 국민순소득 통계를 작성
- **(해석)**    
    - 일반적으로 국민소득 지표는 순액 개념보다는 총액 개념을 주로활용하고 있음.
        - 고정자본소모의 정확한 추계가 현실적으로 어려울 뿐만 아니라 총부가가치에서 고정자본소모가 차지하는 비중이 대체로 안정적이어서 단기간의 시계열을 비교하는 데는 별로 문제가 되지 않기 때문임.
        - 다만, 장기적으로는 기술혁신 등으로 생산구조가 변화하고 고정자본소모도 크게 변동할 가능성이 높기 때문에 장기간의 시계열분석에는 순액 개념이 필요
- **(출처 및 공표일시)**    
    - 한국은행에서는 잠정치 통계발표시 국민총소득(GNI) 통계와 고정자본소모 통계를 발표하며 이를 이용해 국민순소득(NNI)를 계산

#### 5.4 총저축과 총투자
- **(정의 및 지표작성)**    
    - 총저축은 국민처분가능소득(Gross National Disposable Income: GNDI)에서 총소비에 사용되지 않은 부분을 뜻하며, 총투자는 총고정투자와 재고투자를 합한 국네총투자와 국외투자의 합을 의미
        - 국민총가처분소득은 국민총소득에서 국외순수취경상이전을 더한 것으로 이는 최종소비지출과 저축으로 사용
        - 국내총투자(총자본형성)는 고정투자(총고정자본형성)와 재고투자(재고증감)로 나눌 수 있음.
        - 투자를 위한 자금은 대부분 저축에서 조달되며, 저축으로 국내총투자를 충당하고도 남는 부분은 외국에 빌려주고, 모자랄 경우에는 빌려오게 되는데 이를 국외투자라고 부름.

|저축|총저축|국민총처분가능소득 - 최종소비지출|

- **(해석)**    
    - 한 국가의 저축률(투자율)은 인구구성과 경제성장률 등을 포함한 종합적인 경제여건에 의해 좌우
        - 상대적으로 젊은 연령측의 비중이 높거나 개도국같이 높은 경제성장률을 기록한 국가들은 국내 저축을 통한 자본축적이 중요
        - 자본시장이 개방되고 서비스산업이 발전한 내수 중심의 주요 선진국들은 저축 못지않게 소비의 중요성도 높은 편임.
        - 따라서 저축률의 적정수준 여부를 판단하는 것은 쉽지 않음.
