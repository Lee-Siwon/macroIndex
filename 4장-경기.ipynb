# 4장 경기

## 1. 설명

- 생산지수는 일정기간의 생산활동 수준을 나타내는 지표로서 대표적으로 광공업생산지수와 서비스업생산지수 등이 있음.
    - 전산업생산지수는 전체 경기의 흐름과 유사한 움직임을 보이는 동행지표로 GDP와 함께 경기를 판단하는 주요 지표
        - 이와 함께 전산업생산지수 내 광공업 및 서비스업 생산지수를 점검하여 실물 경기 상황을 정확하게 파악
    - 전반적인 경기 동향을 체계적으로 파악하기 위해서 재고, 출하의 움직임도 함께 고려

- 현재의 경기 국면을 파악하고 경기전환점을 정확하게 판단, 예측하기 위하여 동행지수 및 선행지수 순환변동치를 참조
    - 동행지수 순환변동치는 현재 경기 국면 및 전환점 파악에 이용되며, 선행지수 순환변동치는 경기전환점을 예측하는데 활용

- 한편, 기업들의 전반적인 경기체감의 정도를 파악하기 위해서 한국은행과 전국경제인연합회의 기업경기실사지수를 이용

## 2. 생산 

### 2.1 전산업생산지수

- **(정의)**
    - 한국표준산업분류에서 정의된 산업들을 대상으로 재화와 용역에 대한 생산활동을 총합하여 흐름과 변화를 지수로 나타낸 것
        - 산업별지수에는  광공업생산지수, 서비스업생산지수, 건설업생산지수, 공공행정활동지수, 농림어업생산지수

- **(지표작성)**
    - 전산업생산지수는 5개 산업군(광공업, 서비스업, 건설업, 공공행정, 농림어업)으로 이루어져 있음. 
        - 광공업과 서비스업을 제외한 여타 산업의 생산지수는 관련 자료를 기초로 작성
            - 건설업과 공공행정은 각각 건설기성액과 공공부문 지출자료를 기초로하고, 농림어업은 농림수산식품부 '농림어업생산지수'에 통계청의 '어업생산동향조사'를 추가하여 작성
    - 전산업생산지수는 다음의 기준시점 고정 가중평균법(라스파이레스 산식)을 이용하여 계산
        - 가중치는 2015년 기준 GRDP의 산업부문별 총부가가치임.
        $$Q_t = \frac{\sum_{i=1}^n\frac{q_{t,i}}{q_{0,i}}W_{0,i}}{\sum_i^n W_{0,i}}\times 100$$
    $Q$ : 전산업생산지수<br>
    $q$ : 산업별지수<br>
    $W$ : 가중치<br>
    $0$ : 기준시점<br>
    $t$ : 비교시점<br>
    $i$ : 산업 (i=1,...,n)<br>
    
- **(해석)**
    - 	전산업생산지수(Index of All Industry Production; IAIP)는 한국표준산업분류에서 정의된 산업들의 생산활동을 총합하여 지수로 작성한 것으로, 전체 산업생산활동의 단기동향을 파악하고, 주요 정책 수립에 활용되도록 하기위해 작성
    - 전산업생산지수는 모든 재화와 용역의 생산수준을 한 눈에 볼 수 있는 종합지표로 국내경기동향을 파악할 때, GDP와 함께 핵심적인 지표로 사용되고 있음. 
        - GDP는 분기별로 작성되는 반면, 전산업생산지수는 매월 작성되기 때문에 경기동향을 보다 시의성 있게 파악할 수 있음.
        - 산업별로 생산활동을 점검하고, 전산업생산에 미치는 영향을 파악
            - 일반적으로 광공업생산과 서비스업생산의 비중이 크기떄문에 두 산업의 변동이 전체 산업생산에 기여하는 바가 상대적으로 크게 나타남.
            - 다만, 건설업과 공공행정의 경우 비중이 작지만 변동성이 크기때문에 전산업생산에 미치는 영향이 클 수 있으므로 주의할 필요
        - 한편, 전산업생산지수는 현재 2015년(100)을 기준으로 작성되기 때문에, 전산업생산지수가 105인경우 2015년 월평균보다 생산이 5% 증가한것으로 해석
        
- **(출처 및 공표일시)**
    
    - 통계청에서 조사대상월 익월 말일경 잠정치를 발표하며, 보도자료에 해당연도의 월별 공표일정을 함께 제공
        - 전산업생산지수 경로: https://kosis.kr/ ---- 국내통계 / 주제별통계 ---- 경제일반$\cdot$경기 ---- 전산업생산지수 ---- 전산업생산지수(원지수, 계절조정지수)

### 2.2 광공업생산지수

- **(정의)**
    - 일정기간 중 이루어지는 광공업 부문의 생산활동 수준을 나타내는 경제지표의 하나로서, 기준년도(2015년=100)의 생산수준에 비해 현재의 생산수준이 어느 정도인가를 나타내는 대표적인 경기동행지표임. 
    - 우리 경제의 생산 활동 동향을 신속하게 보여주는 지표
        - 광공업생산지수는 기본분류지수와 특수분류지수로 구분되며, 통상은 기본분류지수인 산업별지수를 주로 관찰하되, 특수분류지수에 해당하는 공업구조별지수와 기업규모별지수도 함께 검토

           - 중소기업: 상시종사자 300명 미만인 기업

- **(지표작성)**
    - 통계청은 광공업생산지수 작성을 위해 주요 485개 품목을 생산하는 10,000개 내외의 표본사업체를 대상으로 매월 실시되는 광업ㆍ제조업동향조사를 통하여 생산실적 등을 조사하여 그 결과를 기초로 지수를 작성
        - 우리나라는 연쇄 라스파이레스(Laspeyres) 산식을 이용하여 작성되며, 기준연도에는 2015년 기준 경제총조사 결과 등을 기초로 가중치를 산출하고, 기준년 이후는 매년 광업제조업조사 결과 등을 이용하여 가중치를 산출
        
- **(해석)**
    - 보통 경기가 좋을 때는 공장이 쉼 없이 돌아가고 제품의 생산량은 증가함. 사람들의 소득이 늘면서 제품에 대한 수요가 증가하고 그에 따라 생산 활동이 활발해지기 때문.
    - 광공업생산지수는 한국표준산업분류상 광업, 제조업, 전기ㆍ가스업을 대상으로 매월 작성하는 통계로서 경기 동향을 파악함에 있어서 매우 중요한 경제지표로 이용되고 있으며, 정부의 경제정책 수립이나 민간부문의 투자계획수립 등은 물론 학계나 연구기관에서 장단기 경기변화를 분석하는 경우에도 기초 자료로 사용
    - 광공업 부문의 생산활동을 종합적으로 분석하는데 활용되며, 전반적인 경기 동향을 파악할 때 서비스업생산지수와 함께 핵심적인 지표로 사용됨.
        - 산업별로는 제조업이 광공업생산의 대부분을 차지하고 있으며, 제조업 중에서도 자동차, 반도체및부품, 통신및방송장비, 석유제품, 화학제품, ㅂ차금송, 기타운송장비 등을 주목할 필요
            - 전체 광공업생산에서 자동차와 반도체가 차지하는 비중이 매우 높기 떄문에 자동차제외지수, 반도체제외지수, ICT자동차제외지수 등 일부산업이 제외된 광공업생산지수를 함께 고려
            - 한편, 기본분류 일부항목 제외지수는 총지수(예: 전체 광공업생산에서 특정산업을 제외)외에 광업및제조업, 제조업 등으로 구분되는데 광공업생산에서는 총지수를 선택

### 2.3 서비스업생산지수

- **(정의)**
    - 서비스업의 성장세를 가늠하는 핵심지표로서 서비스업 전체 및 개별업종의 생산 활동을 종합적으로 파악하기 위하여 개별 업종의 상대적 중요도인 부가가치 기준 가중치를 적용하여 지수화한 것.
        - 국가의 경제정책 및 기업의 경영계획 수립을 위한 기초자료, 연구소의 서비스 동향분석 및 연구자료, GDP 추계 자료 등으로 널리 활용

- **(지표작성)**

        - 제조업평균가동률은 경기가 호황일 때는 상승하고, 불황일 때는 하락함.
            - 제조업평균가동률이 지나치게 높으면 공급측면에서 문제가 발생할 소지가 있고 너무 낮으면 생산설비의 효율성이 낮은 수준임을 의미하므로 가동률이 지나치게 높거나 낮은 것은 바람직한 현상으로 보기 어려움.
        - 가동률지수와 광공업생산지수의 변화 방향이 간혹 불일치 하는 경우가 발생하는 데, 이는 주로 가동률지수가 생산량 뿐아니라 생산능력에도 큰 영향을 받기 때문임.
            - 게다가 대표품목 및 대상업종, 가중치 등 통계작성의 기본구조도 상이
        - 업종별로 가동률 수준이 서로 다르고 단순 비교하는 것은 바람직하지 않기 때문에, 통계청에서는 제조업 전체에 대한 평균가동률만 제공하고 있음.

## 4. 경기종합지수

- **(정의)**
    - 경기변동의 국면전환점과 속도, 진폭 등을 측정할 수 있도록 고안된 경기지표로 선행, 동행, 후행종합지수로 구분되어 작성되며, 경기국면 및 전환점을 판단, 예측하는데 동행지수 순환변동치와 선행지수 순환변동치를 참조
        - 선행종합지수는 투자관련 건설수주와 재고순환, 금융 등의 지표처럼 실제 경기순환에 앞서 변동하는 개별지표를 가공$\cdot$종합하여 만든 지수
        - 동행종합지수는 공급측면의 광공업생산지수, 취업자수 등과 수요측면의 소매판매액지수 등과 같이 실제 경기순환과 함게 변동하는 개별지표를 가공$\cdot$종합하여 만든 지수
        - 후행종합지수는 재고, 소비지출 등 실제 경기순환에 후행하여 변동하는 개별지표를 가공$\cdot$종합하여 만든 지수
        - 동행(선행)지수 순환변동치란 동행(선행)종합지수에서 추세변동치를 제거한 것으로 경기 국면과 전환점을 나타내는 지표
            - 경기순환 국면은 경기저점에서 정점까지 경제활동이 활발한 구간을 확장국면, 경기정점에서 저점까지의 경제활동이 위축된 구간을 수축국면으로 구분할 수 있음.
            - 경지저점(정점)에서 다음 저점(정점)까지의 기간을 순환주기라고 하며 정점과 저점간의 차이인 순환진폭은 순환의 강도를 의미

|:---|:---|

- **(지표작성)**
    - 각 부문별로 경기에 민감하게 반응하는 주요 경제지표들을 선정하여 작성
        - 각 지표의 3~4개월 이동 평균한 자료들을 이용하여 증가율을 계산하고, 표준화를 실시한 후 합산한 다음 지수화(기준년도=100)함.
        - 동행(선행)지수 순환변동치는 동행(선행)종합지수에서 국면평균법(PAT법)으로 추출한 추세변동치를 제거하여 산출
        

$$동행(선행)지수순환변동치 = \frac{동행(선행)종합지수}{동행(선행)종합지수의 추세변동치}$$

        
- **(해석)**
    - 동행지수순환변동치는 현재의 경기국면과 전환점을 판단하는데 활용되며, 선행지수순환변동치는 향후경기 국면과 전환점의 예측에 이용
        - 대체로 지표가 2분기 이상 상승하면 경기확장기, 하강이면 경기수축기로 간주하며 가장 높을 때를 경기정점, 가장 낮을 떄를 경기 저점으로 판단
   - 선행지수 순환변동치는 통상적으로 지표가 현재까지와는 반대방향으로 2분기이상 연속하여 움직이면, 경기전환점의 발생신호로 해석
       - 여기에 과거의 평균 선행시차를 더하면 향후 국면전환이 발생할 시점을 대략적으로 추정해 보는 것이 가능

## 5. 기업경기실사지수

- **(정의)**
    - 기업가의 현재 기업경영상황에 대한 판단과 향후 전망등을 나타내는 경제 심리지표로 기업들의 전반적인 경기 체감정도를 파악하는데 활용
        - 기업경기실사지수(BSI)를 작성하는 기관으로는 한국은행, 전국경제인연합회(이하 전경련), 중소기업중앙회, 대한상공회의소 등이 있으며, 조사대상이나조사항목, 조사기간 등에서 다소 차이가 있음.
        - 주로 한국은행BSI나 전경련BSI를 주로 관찰 

- **(지표작성)**
    - 기업 활동의 실적과 계획, 경기 동향 등에 대한 기업가들의 의견을 직접 조사
        - 각 업체의 응답을 아래와 같은 공식에 따라 지수화
        
$$업종별BSI = \frac{(긍정적인 응답업체의 수 - 부정적인 응답업체 수)}{전체 응답업체수}\times 100 + 100$$
$$산업(제조업, 비제조업)별BSI = \sum_i^n w_i\cdot BSI_i$$

**[표 2-4] 한국은행 BSI와 전경련 BSI**

|구분|한국은행 BSI|전경련 BSI|
|:---|:---|:---|
|조사대상|2010년 기준 국세청 법인세 신고기업중 업종별 매출액 누적액 기준 상위 95%업체(전국 2,862개 법인, 제조업 23개업종, 비제조업 13개업종)|전업종별로 전산업기준 매출액 600대기업을 대상(단, 금융업 등 경기대응성이 낮고 조사가 곤란한 일부 업종은 제외)|
|조사항목|업황, 제품재고, 생산설비, 설비투자실행, 인력사정, 신규수주생산매출규모, 가동률, 제품판매가격, 원자재구입가격, 채산성, 자금사정에 대한 당원실적 및 익월 전망|종합경기, 내수판매, 수출 ,투자, 자금사정, 재고, 고용, 채산성에 대한 당월 실적및익월전망|
|조사기간|매월 중순이후 영업일기주 약 6일 조사(매월 다소차이) |매월 중순이후 약 6~10일 조사 (매월 다소차이)|

- **(해석)**
    - 기업경기실사지수는 설문조사 방식이므로, 경기상황에 민감하고 일반적인 경제지표로는 파악하기 어려운 기업활동에 대한 평가와 전망을 수량화 할 수 있다는 장점이 있음.
        - 기준치 100인경우 긍정적인 업체수와 부정적인 업체수가 같고, 100이상이면 긍정적으로 응답한 업체수가 부정적으로 응답한 업체수보다 많음을 의미하고, 100이하인 경우에는 그 반대를 의미함.
            - 우리나라 BSI의 장기평균은 100을 하회하고 있기때문에 장기평균도 함께 고려할 필요
            - 한편, 한국은행 BSI의 조사내용에는 수준과 변화방향에 대한 판단 등이 있기 때문에 계절조정지수의 전월대비 증감을 확인하는 것이 바람직
            - 또한, 각 지수의 세부항목을 확인할 떄에는 각 항목별 조사방법 등을 고려하여 지표를 분석할 필요 
        - 기업경기조사는 조사대상 기업의 경영활동에 대한 주관적인 평가와 전망을 지수화 한 것이기때문에 실물지표와는 변화의 방향이나 정도가 상이할 수 있음.
