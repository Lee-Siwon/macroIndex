## 1.개요

- 국민계정상 민간소비는 가계와 기업의 전반적인 소비활동을 대표하는 지표로서 일반적으로 경기변동과 밀접한 관련을 보임

- 실물경기에 대한 판단을 제고하기 위해서 월별 소비지표를 함께 활용
    
    - 월별 소비관련 지표 중 소비재판매액지수, 소비재 생산 및 출하, 그리고 수입 등은 민간소비의 추이를 파악하는데 유용
    - 다만, 통계청에서 월별로 발표하는 소비 관련 지표와 이에 상응하는 한국은행의 분기별 소비지표는 높은 관련성에도 불구하고 집계방법 등이 상이하여 절대적인 수준 등에 있어서는 다소 차이를 보인다는 점에 유의

- 전반적인 소비여건을 판단하는 과정에서, 민간소비와 밀접하게 관련된 경제지표도 함께 고려
    - 물가상승률과 상용근로자임금, 피용자 보수등 노동소득 관련 자료를 이용
    - 추가적으로 상품군별 경상판매액과 자동차 내수판매, 그리고 해외소비 등의 자료도 참고

## 2.민간소비

### 2.1. 국민계정 민간소비

**(정의)**
- 민간소비지출은 가계최종소비와 가계에 봉사하는 비영리단체소비로 구성
    - 가계의 최종소비지출은 소비를 목적으로 하는 가계가 재화와 서비스에 대해 지출한 금액으로 지출이 발생한 지역과 상관없이 거주자 가계의 지출로 정의
        - 국내에서 이루어진 소비지출에서 거주자의 국외소비를 가산하고 비거주자의 국내소비를 차감하여 산출
    - 가계에 봉사하는 비영리단체의 최종소비지출은 비영리단체가 가계에 대해 재화나 서비스를 무상 또는 낮은 가격으로 제공하는데 투입한 총비용에서 판매수입을 공제

**(지표작성)**
- 민간소비는 최종생산물을 직접 구입하는 가계와 가계에 봉사하는 비영리단체의 지출 현황을 바탕으로 한 수요접근법으로 추계
    - 국민계정상 민간소비는 크게 국내소비와 순해외소비로 구분
        - 국내소비는 거주자의 국내소비는 물론 비거주자의 국내소비를 포함하며, 순해외소비는 거주자의 해외소비에서 비거주자의 국내소비를 차감해서 계산

**(해석)**
- 민간소비(국내소비+순해외소비)는 거주자의 국내외 소비를 의미하는 반면, 국내소비는 소비주체의 국내거주 여부와는 상관없이 국내에서 이루어진 소비활동을 의미
    - 순해외소비가 전체소비에서 차지하는 비중이 작지만, 큰 변동성을 가짐에 따라 민간소비와 국내소비는 상당한 차이를 보일 수 있음.
        - 국제수지상 여행수지(여행수입-여행수출)와 순해외소비(거주자 해외소비-비거주자 국내소비)는 유사한 개념
    - 한편, 환율도 수입물가와 내수용 소비재 수입 등에 영향을 미치고, 결국 민간소비의 변화에 일부 기여하기 때문에 참고지표로 활용

**(출처 및 공표일시)**

### 2.2. 형태별 가계최종소비

**(정의)**
- 형태별 가계최종소비지출은 최종소비재를 크게 재화와 서비스로 구분하며, 재화는 내구성 정도를 기준으로 내구재, 준내구재, 비내구재로 구분

    - 서비스 부문의 국내소비는 협회 등 관련 기관의 기초자료와 생산국민소득의 추계결과 등을 이용

    - 일반적으로 경기가 침체국면일때 내구재와 비내구재 소비의 변화를 비교해 보면, 내구재 소비는 생산보다 감소폭이 크게나타나는 반면, 비내구재 소비는 생산에 비해 감소폭이 작게 나타남.

        - 비내구재 소비(국내소비의 80%)의 변동성이 상대적으로 작게 나타나는 현상은 전체 민간소비의 변동성을 줄이는데 기여

## 3. 소매판매액지수

**(정의)**
- 소매판매액지수는 재화에 대한 소비를 집계하여 지수화한 것으로, 특히 가계부문의 소비재 소비지출 동향을 파악하고 분석하는데 유용
    - 전수 또는 표본조사에 의해 한국표준산업분류상의 자동차 판매업에서 승용차부분과 소매업에 해당하는 업체의 실제 거래된 월간판매액을 2015년 기준(100)으로 지수화
    - 소매판매는 내구재와 준내구재, 그리고 비내구재로 구분되는데, 구체적인 정의는 다음과 같음.

**[표 6-4] 유형별 소비재 구분**

|구분|기준|비고|
|:---|:---|:---|
| 내구재 | 1년 이상 사용가능 & 고가 | 가구, 승용차, 가전제품(TV, PC 등) |
| 준내구재 | 1년 내외 사용 & 저가 | 의류 및 신발, 소형가전 등 |
| 비내구재 | 1년 미만 사용가능 & 저가 | 음식료품, 수도광열, 휘발류, 서적, 문구 등 |

**(지표작성)**
- 통계청은 서비스업동향조사의 약 17,300개의 소매표본사업체조사 결과와 일부 외부기관 자료(수입자동차는 수입자동차협회, 연료는 한국석유공사)를 이용하여 소매판매액지수를 작성
    - 전수조사 대상은 주로 사업체의 수가 적고 판맵중이 큰 업종이며, 나머지 업종은 표본조사를 실시
    - 불변지수는 경상지수를 디플레이터로 나누어 계산하며, 계절조정지수는 불변지수에서 계절변동요인(계정요인, 명절요인 및 영업일수)을 제거하여 작성
        - $$경상지수 = \frac{월별로조사한상품권판매액}{기준년도월평균상품판매액}\times 100$$
        - $$불변지수 = \frac{경상지수}{물가디플레이터}\times 100$$
        - $$계절조정지수 = \frac{불변지수}{명절요인\times 영업일수요인\times 계절요인}\times 100$$
    - 상위업종은 대표업종별로 가중평균(기준년 업종별 부가가치 구성비를 가중치로 적용)하여 산출
    - 한편, 2005년 최초 작성된 소매판매액지수는 2015년 기준으로 개편 단행
        - 기존의 기타 판매액 항목은 지수개편 이후 세분화시켜 내구재, 준내구재, 비내구재의 경상판매금액 기타항목으로 각각 편입

**(해석)**
- 소매판매액지수는 국민계정의 소비재 소비와 유사한 추이를 보이고 시의성도 높기 때문에 전반적인 민간소비 추이를 파악하는데 매우 유용
    - 다만, 통계청에서 월별로 발표되는 소매판매액지수와 한국은행의 분기별 소비지표는 집계방법과 시기에 따른 차이 때문에 분석과정에서 주의할 필요

### 5.2. 기대인플레이션

**(정의)**
- 경제 주체들의 물가에 대한 전망을 인플레이션 기대심리, 또는 기대인플레이션이라고 정의

**(지표작성)**
- 기대인플레이션 측정은 기본적으로 설문조사를 통해 이루어짐.
    - 매월 일주일 동안 한국은행이 전국 56개 도시 2,200여 가구를 대상으로 조사
    - 조사방식은 향후 1년간 물가상승률을 구간별로 나눠 응답자가 생각하는 구간에 표기한후 중앙값으로 산출

**(해석)**
- 기대인플레이션율은 주요 거시경제변수들과 밀접한 관계를 가지고 있기 때문에 소비동향을 파악하는데 참고지표로 활용
    - 기대인플레이션은 다양한 경로로 실물경제에 영향을 미치는 것으로 나타남.
        - 예를 들면, 물가가 오르면서 기대인플레이션이 커지면서 기업이나 자영업자들이 원가부담이 커진 것 보다 가격을 더 올리는 현상이 나타나면서 물가가 빠르게 상승하기도 함.
    - 다만, 기대인플레이션율은 설문조사방법으로 추정하기 때문에 응답자의 주관이 많이 개입되어 신뢰도가 낮은 편임.
