# 7장 설비 투자

## 1. 개요

- 우리 경제의 설비투자 동향은 국민계정상의 설비투자와 더불어 월별지표인 설비투자지수와 국내기계수주, 제조업평균가동률, 자본재 및 기계류 수입액 등을 다각적으로 분석하여 판단
    - 기업경기실사지수중 설비투자 관련 심리지표를 통해 기업의 투자동향과 전망을 파악하는 한편, 주요 기업의 설비투자계획조사 자료도 정기적으로 검토

## 2. 설비투자와 설비투자지수

### 2.1. 국민계정 설비투자

**(정의)**
- 국민계정 총고정자본형성에서 기계류와 운수장비가 설비투자에 해당
    - 기계류는 일반기기, 전기전자기기$\cdot$정밀기기, 기타기계장비, 종축$\cdot$역축$\cdot$낙농축 등으로 구성되며, 운수장비에는 자동차, 항공기, 선박, 철도차량 등이 포함

**(지표작성)**
- 생산된 상품의 처분경로를 따라 투자금액을 추계하는 재화흐름법에 기초하여 작성
    - 설비투자는 국산부문과 수입부문으로 구분됨
    - 국산설비투자는 상품별 출하액에서 통관기준 상품별 수출액을 차감한 국내 총공급에서 상품별 중간소비를 차감하여 국내최종수요를 도출
        - 상품별 출하액은 통계청 광업$\cdot$제조업조사 및 각종 협회자료 등을 통해 계산하고, 상품별 중간소비는 기준년 산업연관표의 국산거래표를 이용하여 산출
    - 수입설비투자는 상품별 수입액에서 중간소비를 차감한 수입 최종수요에 산업연관표상에서의 소비, 고정투자, 재고투자 비율을 적용하여 추계
        - 상품별 수입액은 관세청 통관 통계를 이용하고, 상품별 중간소비는 산업연관표 수입거래표를 이용해 산출
    - 국산 및 수입 설비투자에 부대비용 및 각종 마진을 합산하여 최종적인 설비투자 금액을 산출

**(해석)**
- 설비투자는 수요와 생산능력을 확대시키는 효과가 있기 때문에 경기 동향을 파악하는데 매우 중요한 지표 
    - 설비투자는 단기적으로 경제의 총수요 증가에 기여하고, 장기적으로는 경제의 생산능력 증대를 통해 경제성장과 생산구조 조정에 영향을 미침.

**(출처 및 공표일시)**
- 한국은행은 1, 4, 7, 10월 25일경에 전분기 국내총생산 속보치를 발표하고 3, 6, 9, 12월에 잠정치를 발표하며 다음해 3월말에 확정치를 발표
    - 한국은행 경제통계시스템 홈페이지에서 공표일정을 제공하고 있음.
    - 설비투자는 한국은행의 국민계정 또는 분기 국내총생산 보도자료에 포함되며, 경제통계시스템에서 관련 자료를 모두 제공
        - 자본재 형태별 설비투자 경로: https://ecos.bok.or.kr/ ---- 통계검색 ---- 2. 국민계정 ---- 2.1. 국민소득통계(2015년 기준년) ---- 2.1.6. 총자본형성 ---- 2.1.6.1. 자본재형태별 총자본형성 ---- 2.1.6.1.1~4 자본재형태별 총자본형성 ---- 총자본형성 ---- 설비투자(운송장비, 기계류)

### 2.2. 설비투자지수

**(정의)**
- 자동차, 항공기, 철도차량, 선박 등의 운수장비와 제조업 등에서 생산설비로 사용되는 각종 기계류 등 설비투자재의 공급 및 배분 과정을 추적하여 매월 설비투자액을 추정한 후 지수화하여 집계한 월별지표
    - 설비투자란 생산과정에서 1년이상 반복적 또는 지속적으로 사용되는 자산, 즉 제조업 등에서 생산설비로 사용되는 각종 기계류와 자동차, 항공기, 선박 등의  운송장비를 구입하거나 취득하는 행위를 말함. 통계청 설비투자지수는 사업체에서 생산한 설비투자재의 투자액을 파악하는 공급 측면의 가공통계로서 개량상품흐름법에 근거하여 작성.
    
    - 2015년 기준 산업연관표의 총 381개 품목 중 고정자본형성액이 있는 기본부문은 68개이며, 이 중 3개 부문을 제외한 65개 부문을 선정
        - 화훼작물, 낙농, 기타축산, 영림 등 4개 부문은 지수 작성에 필요한 기초자료 확보가 곤란하고 그 비중이 미미하여 제외
        -  65개 기본부문에 부속되는 품목은 444개로 설비투자지수에서는 이 444개의 품목을 지수 작성대상인 대표품목으로 하고 있음
        
    - ICT(Information and Communication Technology)산업이 우리 경제에 미치는 영향이 꾸준히 증가함에 따라 OECD의 ICT 산업분류에 해당하는 컴퓨터 및 주변기기 등 7개 부문을 대상으로 ICT 설비투자지수를 작성
    - 2015년 기준으로 설비투자지수를 개편하면서 추계모형 변경, 선박 등 작성범위 확대, 선박 및 항공기 제외지수 작성, 최근 지표 포함 등을 총해 산업구조변화를 반영하고 여타 통계들과의 적합성을 제고

**(지표작성)**
- 설비투자지수는 생산자로부터 최종수요자에게 이르기가지의 상품 유통과정을 반영하여 간접적으로 지출금액을 추계하는 개량상품흐름법에 따라 작성
    - 설비투자재의 총공급(내수+수입)에서 중간수요와 최종수요(소비)를 제외한 나머지가 고정자본으로 배분된다고 가정
    - 매월 불변국내생산액, 불변수입액, 불변수출액, 불변재고증감액, 불변중간수요액, 불변설비투자액을 추계한 후 기준년 평균금액 기준으로 지수화 (*t*: 당해, *0*: 기준)
        - $$설비투자지수(t)=\frac{국산설비투자액(t)+수입설비투자액(t)}{기준년월평균설비투자액(0)}\times 100$$

**(해석)**
- 설비투자지수는 월별지표이기떄문에 국내 설비투자의 동향을 신속하게 파악하는데 매우 유용함
    - 다만, 설비투자지수는 국민계정 설비투자와 달리 화훼작물, 낙농등이 제외되어 있고, 자료 수집의 제약 등으로 작성과정이 다소 상이함에 유의
- 한편, 향후 설비투자의 변화 추이를 예측하기 위해 설비투자조정압력(제조업생산증가율과 생산능력증가율의 차이)도 함께 점검
    - 생산능력증가율이란 제조업 생산능력지수의 전년동기대비 증가율로서, 이때 생산능력이란 사업체가 정상적인 조업환경 하에서 생산할 경우의 최대 생산가능량을 의미

**(출처 및 공표일시)**
- 통계청에서 매월 말일이나 마지막 주 금요일 전월의 잠정치를 발표하고 2개월 후 확정치를 발표
    - 설비투자지수는 산업홀동동향 보도자료에 첨부되고, 국가통계포털의 주제별통계에서도 관련 자료를 함께 제공
        - 설비투자지수 경로: https://kosis.kr ---- 국내통계/주제별통계 ---- 경제일반$\cdot$경기 ---- 설비투자지수

### 2.3. 설비용기계류 내수출하지수

**(정의)**
- 국내 판매업자, 타사업체, 기관, 단체 및 개인소비자 등에 판매된 국산 설비용기계류의 추이를 파악할 수 있는 지수로 생산자제품출하지수 대상품목에서 설비용기계류에 해당되는 97개 품목으로 작성 
    - 설비용기계류는 한국표준산업분류의 금속가공(분류코드:26~31)까지의 중분류 중 기계설비 및 이의 구성품에 해당

**(지표작성)**
- 2015년 기계류 품목볗 월평균 물량을 기준물량으로 사용하여 월별 비교물량을 기로준물량으로 나누어 품목별 지수를 작성한 뒤 작성된 품목별 지수를 가중평균함.
    - $$기계류내수출하지수 = \sum \frac{당월물량}{기준물량}\times 가중치$$

**(해석)**
- 일반적으로 출하가 생산보다 경기에 민감하게 반응하므로 경기동향을 판단하는 과정에서 설비투자지수와 함께 종합적으로 분석
    - 경기 하강(상승)이 시작되는 경우, 출하는 감소(증가)하지만 생산은 바로 조정되지 않는 경우가 많은데, 이런 경우 재고가 증가(감소)하므로 경기 동향의 체계적인 판단을 위해 관련 지수간의 움직임을 종합적으로 검토할 필요
        - 다만 이들 지수간에는 조사대상 및 조사기간, 가중치 등의 차이로 예상했던 관계가 항상 성립하지 않을 수 있음을 유의

**(출처 및 공표일시)**
- 통계청에서 매월말이나 마지막 주 금요일에 전월의 잠정치를 발표하고, 2개월 후 확정치를 발표
    - 기계류내수출하는 산업활동동향 보도자료에 포함되고 국가통계포털의 주제별통계에서도 관련 자료를 제공
         - 기계류내수출하지수 경로: https://kosis.kr ---- 국내통계/주제별통계 ---- 광업$\cdot$제조업 ---- 광업제조업동향조사 ---- 생산,출하,재고 ---- 설비용기계류생산지수

## 3. 내수용자본재수입액과 기계류수입액

### 3.1. 내수용자본재수입액

**(정의)**
- 1년 이상 반복적, 지속적으로 사용되는 생산기계나 생산수단을 만들어내는 제품의 수입액중 내수용 제품의 금액
    - 설비투자에 직접적으로 이용되는 장비부문과 부품으로 구분 가능
        - 공구, 기계, 장비, 부품, 컴퓨터, 의료용 기기, 계측기, 반도체, 건전지, 센서, 발전기, 전동기, 전기부품 등 

**(지표작성)**

## 4. 국내기계수주

**(정의)**
- 설비용기계류를 생산하는 주요 제조업체의 수주액을 수요자 및 기계종류별로 매월 조사
    - 수요자는 공공과 민간으로 나뉘고, 민간수요에서 제조업과 비제조업으로 분류
    - 기계종류는 원동기, 특수산업용기계, 금속공작$\cdot$가공기계, 일반산업용기계, 도로주행차량, 기타수송용기계, 의료$\cdot$w정밀측정제어기기, 철구조물 등으로 분류 

**(지표작성)**
- 조사대상 기간(매월 1일~말일) 중 주문 받은 총 계약금액으로 표시됨.
    - 전전년 광업$\cdot$제조업 조사결과의 종업원 수 및 기계종류별 생산액 등을 고려하여 총생산액의 65%에 해당하는 상위 기업체를 대상으로 조사
    - 외부의 주문 없이 업체의 계획에 따라 생산한 계획생산과 제조업체가 기업 안에 설치하여 사용할 목적으로 생산한 자가사용분이 포함.

**(해석)**
- 국내기계수주는 약 1~3개월후의 국내설비투자로 연결되므로 향후 설비투자 규모를 예측하는데 활용되는 선행지표임.

**(출처 및 공표일시)**
- 통계청에서 매월 말일이나 마지막 주 금요일에 전월의 잠정치를 발표하고, 2개월 후 확정치를 발표
    - 국내기계수주 경로: https://kosis.kr ---- 국내통계/주제별통계 ---- 광업$\cdot$제조업 ---- 기계수주동향조사 ---- 수요자별 기계수주(불변금액, 계절조정)

## 5. 설비투자 기업경기실사지수

### 5.1. 한국은행 기업경기실사지수

**(정의)**
- 현재 설비투자 상황에 대한 기업가의 판단과 향후 전망을 조사하여 지수화 한것으로 전반적인 설비투자여건을 판단하는데 유용

**(지표작성)**
- 매월 15일에서 23일간 전국 약 2,800여개의 법인기업 중 제조업체를 대상으로 설비투자의 실행 수준을 조사하여 지수화
    - 연매출액 기준 해당 업종 상위 15% 이내 영리법인업체로 제조업이 60~65%를 차지
    - $$설비투자BSI = \frac{계획대비 수정증액 응답업체수-계획대비 수정감액 응답업체수}{전체응답 업체수}\times 100 + 100$$

**(해석)**
- BSI가 100을 넘어서면 설비투자를 당초 계획보다 증가시켰다는 응답업체가 감소시킨 업체보다 많음을 의미
    - 기업의 관점에서 설비투자에 대한 현재 경기판단이나 전망이 긍정적 혹은 부정적인지를 파악하는데 이용

**(출처 및 공표일시)**
- 한국은행에서 매월 말일경에 당월의 기업경기실사지수를 보도자료를 통해 공표
    - 설비투자 기업경기실사지수 경로:  https://ecos.bok.or.kr/ ---- 6. 심리지수 ---- 6.1. 기업경기실사지수 ---- 6.1.1. 기업경기실사지수 ---- 6.1.1.1.(실적), 6.1.1.2.(전망)

### 5.2. 전경련 기업경기실사지수

**(정의)**
-   전경련 투자 BSI 실적은 지난달 대비 이번달 투자의 증감여부를 지수로 나타내며, 투자 BSI 전망은 이번달 대비 다음달 투자계획 증감 여부를 지수로 표현

**(지표작성)**
-   매월 중순에 약 5일간 매출액 기준 상위 600 기업을 대상으로 응답기업 담당자의 자기기술과 조사원의 질의 기술을 병행하여 작성
    - $$투자BSI = \frac{긍정응답기업수-부정응답기업수}{전체응답기업수}\times 100 +100$$

**(해석)**
-   BSI가 100을 넘으면 설비투자가 전월보다 증가한 응답업체가 감소한 업체보다 많음을 의미
    -  기업의 관점에서 설비투자에 대한 현재 경기판단이나 전망이 긍정적 혹은 부정적인지를 파악하는데 활용
    - 한편, 전경련 투자 BSI는 상대적으로 기업규모(매출액기준)가 작은 업체가 제외된다는 점에서 한국은행의 자료와 다소 상이

**(출처 및 공표일시)**
-   전국경제인연합회에서 매월 말일경 보도자료 형태로 공표
    - 전경련 기업경기실사지수 경로: https://www.fki.or.kr/ ---- 전경련소식 ----보도자료 ---- BSI 검색

## 6. 설비투자전망

**(정의)**
- 국내 설비투자의 연구 및 정부의 정책수립에 필요한 기초자료를 제공하기 위해서 국내 주요 기업의 설비투자실적과 계획을 조사
    - 세부항목으로 기업규모별, 업종별, 투자동기별, 자금조달별 설비투자계획을 조사하며 설비투자의 결정 및 부진요인 등을 분석

**(지표작성)**
- 전국 약 2,200개(상반기), 3,500개(하반기) 이상의 주요 사업체를 대상으로 6, 12월에 걸쳐 전년도 설비투자 실적금액 및 당해년도 설비투자 계획금액을 조사하되 설문지와 조사원위 구술조사 방법을 병행
    - 대략적으로 대기업 20%, 중견기업 35%, 중소기업 45%로 구성됨

**(해석)**
- 설비투자에 대한 전망과 세부항목을 이용한 미시적 접근 및 분석에 활용
    - 각 기업규모별, 투자동기별, 자금조달별 설비투자의 실적 및 계획을 발표하므로 일반적인 거시지표에서 제외된 설비투자 관련 세부사항을 파악할 수 있음.
