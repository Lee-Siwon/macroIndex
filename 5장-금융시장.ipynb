# 5장 금융시장

## 1.개요

- 국내 금융시장의 주요 가격변수와 자금 흐름의 변화를 점검하여 전반적인 금융시장 여건을 신속히 파악
    - 대내외 불확실성이 확대되고 국내외 주요 금융지표들의 동조화 현상이 심화되는 추세임을 감안할 때 금리, 주가, 환율 등의 추이를 지속적으로 관찰할 필요
    - 금융기관 여수신, 통화량 등을 통해 국내 금융시당의 자금 흐름도 함께 파악
    - 한편, 금리와 환율 등 일부 지표에 대한 정의와 해석은 추후 세계경제의 국제금융시장 부문에서 다룰 예정

## 2. 금리

### 2.1. 기준금리

- **(정의 및 지표작성)**
    - 한국은행이 금융기관과 환매조건부증권(RP) 매매, 자금조정 예금 및 대출 등의 거래를 할 때 기준이 되는 정책금리
        - 한국은행은 기준금리를 7일물 RP매각시 고정입찰금리로, 7일물 RP매입시 최저입찰금리(minimum bid rate)로 사용, 자금조정 예금 및 대출금리에서 각각 -100bp와 +100bp를 가감하여 운용
        - 금융통화위원회는 물가 동향, 국내외 경제 상황, 금융시장 여건 등을 종합적으로 고려하여 연 8회 기준금리를 결정
        
- **(해석)**
    - 기준금리를 조정하면 자금시장에 영향을 미쳐 초단기금리인 콜금리에 즉시 영향을 미치고, 장단기 시장금리, 예금 및 대출금리 등의 변동으로 이어져 궁극적으로는 투자와 소비 등을 통해 실물경제 활동에 영향

### 2.2. 시장금리

- **(정의 및 지표작성)**
    - 금리는 금융시장에서 자금을 빌린 대가로 지불하는 사용료를 원금으로 나눈 비율이며, 일반적으로 자금에 대한 수요, 공급의 변화에 따라 시장에서 결정
        - 일별 장단기 금리를 집계하여 함께 비교함으로써 매월 변동하는 시장금리 추이를 조사
        - 콜금리, CD금리, CP금리, 국고채 3년물 수익률, AA-등급 및 BBB-등급 3년물의 회사채수익률 등의 자료를 점검
        
- **(해석)**
    - 금리는 가계의 저축, 기업가의 투자활동, 물가수준, 국가간의 자금흐름, 경기변동 등 경제 전반에 광범위한 영향을 미치기 때문에 움직임을 주의 깊게 관찰할 필요
        - 일반적으로 단기금리는 중앙은행의 통화정책 운용방향, 인플레이션 기대 등을 반영하여 수시로 변동되고, 장기금리는 현재의 단기금리, 미래 단기금리에 대한 시장의 기대, 기간프리미엄(term premium) 등에 따라 결정
        - 장단기 금리차는 향후 경기전망 및 인플레이션에 대한 시장의 기대를 반영
            - 향후 경기가 좋아질 것으로 기대하는 경우 인플레이션이 현 수준보다 상승할 것으로 예상하여 장단기금리의 격차가 확대되는 반면, 경기 침체로 미래의 단기금리가 현재 수준보다 낮아질 것으로 예상되는 경우 장단기금리차가 축소되거나 간혹 금리가 역전되기도 함.

### 2.3. 채권에 반영된 기대인플레이션

- **(정의 및 지표작성)**
    - 국고채(10년)와 물가연동국채(10년)간의 금리차이를 이용하여 채권가격에 반영된 기대인플레이션을 계산
        - 물가연동국채는 채권의 원금과 이자를 물가에 연동시켜 국채 투자에 따른 물가변동위험을 제거하여 채권의 실질구매력을 보장한 국채
        
- **(해석)**
    - 국고채와 물가연동국채 간의 금리차는 투자자들의 기대인플레이션의 대용지표로 간주되어 통화정책 수행에 유용한 정보로 활용될 수 있음.
        - 채권에 반영된 기대인플레이션은 향후 경기와 물가를 전망하는 보조적 지표로, 두 채권 간의 금리차가 커질수록 기대인플레이션이 증가
        
- **(출처 및 공표일시)**

## 3. 주가

### 3.1. 주가지수

- **(정의)**
    - 주식시장 전체의 가격움직임을 파악하기 위해 개별주식의 주가변동을 종합하여 작성한 지수
        - 우리나라에서 발표되고 있는 주가지수느 대표적으로 종합주가지수와 코스닥지수 등이 있으며 모두 시가총액식 주가지수 산정방식을 채택

- **(지표작성)**
    - 주가지수는 기준시점과 비교시점의 주가수준을 비교하여 산출
        - 종합주가지수(KOSPI)는 유가증권시장에 산장된 전 종목을 대상으로 개별종목의 주가에 상장주식수를 가중한 기준시점(1980년 1월 4일 = 100)의 시가총액과 비교시점의 시가총액을 대비하여 산출
        - 코스닥지수는 코스닥시장에 산장된 기업의 주식가격에 주식수를 가중평균한 시가총액지수(2004월1월26일=1,000)

$$시가총액방식주가지수 = \frac{비교시점의 주식 시가총액}{기준시점의 주식 시가총액}\times 100$$
        
- **(해석)**
    - 주가지수에는 다양한 경제적 정보가 반영되지만, 경기상황 판단의 보조지표라는 관점에서 주로 주가지수의 변화를 초래한 요인들에 주목하는 것이 바람직
        - 주가지수는 기업의 영업성과 및 자금조달비용, 투자자의 실물경기에 대한 판단 등 다양한 정보가 내재된 변수이기 떄문에 실물경기 판단에 유용 
        
- **(출처 및 공표일시)**
    - 일자별 및 시간대별 주가는 한국거래소에서 확인 가능
        - 한국거래소 주가지수 경로: http://data.krx.co.kr/contents/MDC/MAIN/main/index.cmd ---- 지수 ----주가지수 ---- 개별지수 종합정보 --- 일자별 시세
        
- **(조사주기 및 자료의 형태)**
    - 한국거래소는 주가지수를 10초단위로 발표하고 있으며, 1980년 이후의 종합주가지수(KOSPI)와 1996년 7월 이후의 코스닥지수(KOSDAQ)를 일별, 월별, 연도별로 제공
        - 시가 및 종가 지수, 등락률, 거래량, 거래대금 등을 확인할 수 있음.
        - 또한, 투자자들을 위해 1975년 1월 4일까지의 과거 지수를 소급하여 발표하고 있음

### 3.2. 투자자예탁금

- **(정의)**
    - 투자자가 유가증권의 매매거래 등과 관련하여 증권회사에서 일시적으로 맡겨놓은 예수금을 의미
        - 위탁자 예수금, 청약자 예수금, 저축자 예수금, 환매조건부 예수금, 신용거래 구좌설정 보증금, 신용거래 보증금 등으로 구성

- **(해석)**
    - 일반적으로 투자자 예탁금은 언제든지 주식매입 자금으로 동원될 수 있기 때문에 투자자 예탁금이 증가하면 주가 상승을 기대하는 투자자들이 많다는 것을 시사 
        -  다만, 주식을 사기 위해 적립한 금액 뿐아니라 주식을 팔 경우에 일시적으로 입금되는 회수금도 포함되어 있기 때문에 투자자 예탁금만으로 주가지수를 예상하기는 어려움
        - 개인투자자들의 투자동향을 반영하는 지표인 만큼, 주식시장을 판단하는 보조지표로 활용하는 것이 바람직 
        
- **(출처 및 공표일시)**
    - 금융투자협회에서 일별 투자자예탁금 수치를 확인할 수 있음.
        - 투자자예탁금 경로: https://freesis.kofia.or.kr/ ---- 주식 ---- 증시자금추이 ---- 투자자예탁금
        
- **(조사주기 및 자료의 형태)**

### 3.3. 외국인 주식투자

- **(정의 및 지표작성)**
    - 금융감독원에 외국인 투자자로 등록한 주체들의 주식투자를 의미하며 투자 등록한 외국인의 보유주식을 조사
        - 한편, 우리나라는 자본시장 국제화의 일환으로 1992년 국내 주식시장을 개방한 바 있음. 

- **(해석)**
    - 증권시장의 3대 주체(외국인, 기관, 개인)중 하나로 외국인의 국내 증시 영향력이 매우 크므로 외국인 투자의 변화 추이를 파악하는 것은 매우 중요
        - 대규모 해외자금의 유출입은 채권시장, 외환시장 등 국내 금융시장 전반에 큰 영향을 미침.
        - 외국인 투자를 결정하는 요인으로는 시장규모, 유동성, 신용등급, 재정건전성, 통화가치, 실물경제여건 등이 있음.
            - 증권시장에 참여하고 있는 외국인들의 경우 다양한 투자주체가 존재하는 점을 감안할 때, 외국인 투자의 근거를 명확하게 설명하는 것은 어려움
            - 일반적으로 국제 주식투자는 투자 대상국의 주가, 환율 등 투자수익요인과 투자국의 금리, 위험 등 투자비용요인, 국가간 상대수익률 및 위험등을 종합적으로 고려하여 결정
        -우리나라는 신용위험, 경제 여건이 비교적 양호한 편이지만, 세계경제 위기시 자금 이탈에 취약하므로 외국인 투자 동향을 점검하여 적절하게 대응할 필요

- **(해석)**

## 5. 통화

### 5.1. 통화량

- **(정의 및 지표작성)**
    - 통화량은 한 나라의 경제에서 일정 시점에 유통되고 있는 화폐의 총량으로 협의통화(M1), 광의통화(M2), 금융기관 유동성(Lf) 등으로 구분
        - 2002년부터 IMF의 통화금융통계매뉴얼(2000년) 기준에 따라 지표를 편제하여 공표
            - 금융기관 분류기준을 바탕으로 예금취급기관(중앙은행 포함)의 연결대차대조표를 이용하여 M1, M2를 편제

**[표 5-3] 통화 및 유동성 지표의 정의와 구성**

|구분|정의|구성|
|:---|:---|:---|
| 본원통화 | 통화량의 기초를 이루는 자금원천 | 민간화폐보유액, 금융기관지급준비금 |        
| 협의통화 (M1) | 회폐의 지급결제수단 기능을 중시한 지표로 단기금융 시장의 유동성 수준을 파악하는데 적합 | 현금, 요구불예금, 수시입출금예금 |        
| 광의통화 (M2) | 협의통화보다 넓은 의미의 지표로 유동성이 높은 항목이 추가되어 시중의 통화량을 파악하는데 용이 | M1, 정기예적금, 시장형 금융상품, 실적배당형 금융상품, 기타예금, 금융채 등 |        
| 금융기관 유동성 (Lf) | 금융기관이 공급하고 있는 유동성만을 포괄 | M2, 2년이상장기금융상품, 생명보험계약준비금 및 증권금융예수금 등 |        

- **(해석)**
    - 경제가 안정적으로 성장하기 위해서는 통화의 양이 적정수준으로 유지되어야 하며, 금리를 중시하는 통화정책의 운용체제로 변경됨에 따라 통화량 지표는 정책운용의 핵심 정보변수로 활용되고 있음.
        - 협의통화는 상대적으로 한국은행의 정책이 직접적으로 반영되며, 광의통화는 경제 전반에 대한 파급효과를 점검하는데 활용
            - 특히, 협의통화화 광의통화의 움직임이 상이할 때 그 원인을 확인
        - 통화량의 적정성을 평가하기 위해서는 일반적으로 명목GDP 대비 통화량 비중 및 통화승수(M2/본원통화) 등을 함께 검토

## 7. 금융기관 여신 및 자산운용

        - 은행의 기업대출은 대기업, 중소기업 대출로 구분하고 CP, 회사채, 주식발행을 통한 자금조달 추이도 함께 확인

    - 신규분양 아파트의 입주예정자를 대상으로 시공사가 별도의 대출 협약을 통해 연대 보증하는 집단대출도 함께 검토

| CP순발행 | 주로 단기자금 조달을 위해 기업이 발행하는 어음형식의 단기 채권으로 신용도가 높은 우량기업이 자기신용으로 발행 | 종금사, 증권사 및 은행신탁 취급분|

| 주식발행 | 주식회사 설립자본금을 조달하거나 자본금을 증액하기 위하여 발행 | 증권신고서 제출(공모)기준, 납입일 기준 신주모집통한 자금조달 실적 |

- **(해석)**
    - 금융기관의 여신 추이를 살펴보고, 급격한 변화가 발행했을때 그 원인을 파악
        - 은행대출(간접금융)과 주식 및 채권발행(직접금융)의 기초자료를 통해 기업의 자금 사정을 파악
        - 대기업의 은행대출 증가요인으로 사업투자 및 M&A 자금수요 등을 고려해 볼 수 있음
        - 중소기업은 주로 은행대출에 의존하는 등 직접금융시장을 통한 자금조달 규모가 작은 편임.
        
- **(출처 및 공표일시)**

## 8. 연체율

- **(정의 및 지표작성)**
    - 대출의 부실화와 은행의 건전성 등을 파악하기 위해 국내은행의 원화대출(은행, 신탁계정의 기업대출 및 가계대출)의 연체율 추이를 관찰
        - 기업별(대기업/중소기업), 가계대출별(주택담보대출 포함) 연체율의 시계열 자료를 비교
        - 건설업, 선박건조업 등 일부 취약업종별 기업대출의 연체율 추이도 함께 점검
        - 일반적으로 원금이 약정기일에 상환되지 않으면 익일부터 연체채권으로 계상되는 1일 이상 원금 연체기준으로 산정되며, 자세한 내용은 보도자료를 참고

$$연체율 = \frac{연체대출채권}{총 대출채권}\times 100$$
        
- **(해석)**
    - 은행의 중요한 보유자산인 대출대권의 질적 수준을 평가하는 대표적인 건전성 지표로서 ,연체율 추이와 상승 및 하락의 원인을 확인하는 것이 중요
        - 연체율을 통해 금융기관의 건전성 및 수익성 등을 점검하여 국내 금융시장여건을 평가
        - 한편, 가계와 기업의 연체율 추이를 비교하여 경제주체별, 산업별 실물경기여건도 함께 점검
        
- **(출처 및 공표일시)**
    - 금융감독원 은행 감독국에서는 매월 말 국내은행의 전월 대출채권 및 연체율 현황의 잠정치를 발표
        - 연체율 경로: https://www.fss.or.kr/ ----  보도,알림 ---- 보도자료 ---- 연체율검색
        
- **(조사주기 및 자료의 형태)**
    - 금융융감독원은 2007년부터 매월 국내은행의 대출채권 연체율 현황을 공표
        - 금액기준(말잔) 연체율과 계절조정 연체율을 보조지표로 매월 발표
        -한국은행 경제통계시스템에서도 연체율 자료를 엑셀로 다운로드할 수 있으나, 소수점 1자리까지만 제공

## 9. 가계신용

- **(정의 및 지표작성)**
    - 일반 가정이 은행등 금융기관에서 대출을 받거나 외상으로 물품을 구입한 대금 등을 합한 금액으로 가계대출과 판매신용으로 구성
        - 가계대출은 금융기관 대출을 포괄하며 예금은행, 비은행 예금취급기관, 기타 금융기관 등으로 구분
        - 판매신용은 재화의 판매(생산)자나 서비스 제공자가 제공하는 외상(신용)거래
            - 신용카드사, 할부금융사 등의 여신전문기솬 및 백화점, 자동차회사, 가전회사등 판매회사가 제공하는 외상거래
        - 2011년 8월 한국은행은 비은행권에 증권사, 대부사업자, 연기금 등을 포함하는 등 종전의 가계신용체제를 개편한 바 있음.
        
- **(해석)**
    - 가계부채의 수준과 변화율 등을 확인하여 금융기관의 건전성을 파악하는 한편, 향후 민간소비 전망에도 활용
        - 경제규모의 확대와 금융시장의 발전으로 자금중개기능이 제고됨에 따라 가계부문의 금융자산과 부채가 함께 증가하는 추세임.
            - 가계부문의 부채 증가를 단순히 부정적으로 해석하기 보다는 자산측면을 함께 고려하여 가계의 재정적 건전성을 판단하는 것이 바람직
        
- **(출처 및 공표일시)**
    - 한국은행에서는 분기별로 전분기의 가계신용 잠정치를 발표하며, 경제통계시스템에서 관련자료를 함께 제공
        - 가계신용 경로: https://ecos.bok.or.kr/ ----  통계검색 ---- 1.통화/금융 ---- 1.2. 금융기관 여수신 ---- 1.2.4.가계신용 ---- 세부계정 선택
        
- **(조사주기 및 자료의 형태)**
    - 한국은행에서는 1997년부터 분기별 가계신용 동향을 제공하며, 월별 예금 취급기관의 가계대출 자료는 월별 금융시장 동향을 통해 발표
        - 두 자료 모두 경제통계시스템에서 엑셀로 다운로드 가능
