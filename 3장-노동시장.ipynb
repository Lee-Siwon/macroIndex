# 3장 노동시장

## 1. 개요

- 노동시장의 경제지표는 크게 `고용`과 `임금` 부문으로 분류되며, 각각 *경제활동인구조사* 와 *사업체노동력조사*를 기초로 작성되고 있음.
    - 고용부문에서는 전년동기대비 취업자수 증감, 실업률, 고용률 지표등을 점검하여, 전반적인 국내고용상황을 평가
        - 특히, 연령별, 산업별, 종사상지위별, 사업체규모별 등으로 구분하여 각 지표들의 변화 추이를 면밀하게 분석
    - 임금부문에서는 근로자의 급여와 근로시간을 정기적으로 확인하고, 종사상지위별 및 산업별 등 세부항목으로 구분하여 근로실태를 평가

## 2. 고용 :  경제활동인구조사

- **(정의)**
    - 경제활동인구조사는 국민 경제활동(취업, 실업, 노동력 등)의 특성을 조사하여 거시 경제관련 분석과 인력자원의 개발정책 수립등에 필요한 기초자료를 제공

- **(지표작성)**
    - 매월 15일이 포함된 1주간(일~토)을 조사대상 기간으로 설정하고, 그 다음주에 조사를 실시
        - 표본규모는 약 33,000가구이며, 조사대상은 표본가구에 상주하는 자 중에 만 15세 이상인 자
            - 지방통계청 사무소에서 담당직원이 표본조사가구를 방문하여 면접조사(PDA 혹은 종이조사표 이용)하거나, 전화조사(CATI) 혹은 인터넷조사(CASI)를 샹황에 맞게 실시
        
- **(해석)**
    - 매월 대규모의 표본을 조사하기 떄문에 정확하고 신속한 통계를 제공하지만, 일부 한계점도 존재
        - 매월 제공되는 고용지표를 기초로 미시적인 분석이 가능
        - 원칙적으로 외국인도 조사대상에 포함되지만, 외국인의 응답수가 극히 적어서 유의성이 떨어지기 떄문에 외국인과 내국인을 구분하는 변수는 공개되지 않고 있음.
        - 매월 조사기간에 설날이나 추석과 같은 명절이 포함되면 실업률이나 고용률 등의 지표들이 급등락한다는 사실에 주의할 필요
        
- **(출처 및 공표일시)**
    - 매월 15일경 통계청에서 보도자료와 함께 발표

    - 통계청 보도자료에 1년동안 월별 통계공표 일정을 공지
    -  통계청과 고용노동부 홈페이지에서 동일한 내용의 보도자료를 발표하며, 당일 기획재정부에서는 독자적으로 분석자료를 발표

### 2.1. 생산가능인구와 경제활동인구

-  **(정의)**
    - 총인구중 만 15세 이상의 인구를 *생산가능인구* 라고 하며, 이들 가운데 일하고자하는 의지가 있는 사람들을 *경제활동인구*, 그리고 경제활동인구에 포함되지 않는 사람을 *비경제활동인구* 라고 정의
        - 경제활동인구는 조사대상 기간동안 상품이나 서비스를 생산하기 위하여 실제로 수입이 있는 일을 한 *취업자* 외에 일을 하지는 않았으나 구직활동을 한 *실업자* 도 포함됨.
-  **(지표작성)**
    - 경제활동인구조사의 표본 정의상 조사에 은한 모든 대상이 생산가능인구에 해당
        - 생산가능인구 = 경제활동인구 + 비경제활동인구
        - 경제활동인구 = 취업자 + 실업자

### 2.2. 취업자

- **(정의)**
    - 취업자는 (1) 조사대상주간에 수입을 목적으로 1시간 이상 일한 자, (2) 동일가구내 가족이 운영하는 농장이나 사업체의 수입을 위하여 주당 18시간 이상 일한 무급가족종사자, (3) 직업 또는 사업체를 가지고 있으나 일시적인 병 또는 사고, 연가, 교육, 노사분규 등의 사유로 일하지 못한 일시휴직자를 포함
    
- **(지표작성)**
    - 취업자 여부를 결정하는데 자세한 과정과 내용은 위의 조사표 문항을 참고
        - 복잡한 논리적 판단을 하지 않고서 간단하게 취업자 여부를 판단할 수 있음.
        - 조사표 23번 문항의 "지난 4주내에 직장(일)을 구해 보았습니까?"의 값이 0 이면 취업자

### 2.3. 실업자

- **(정의)**
    - 조사대상주간에 수입 있는 일을 하지 않았고, 지난 4주간 일자리를 찾아 적극적으로 구직활동을 하였던 사람으로서 일자리가 주어지면 즉시 취업이 가능한 자
    - 실업자의 정의는 국가별로 상이하며 ILO기준과 OECD기준으로 나누어 볼 수 있는데, 두 기준의 큰 차이점은 구직활동의 기간과 연령임.
        - OECD는 구직활동 기간을 4주간으로 설정하고 연령을 만 15~64세로 한정하고 있는 반면, ILO는 구직활동과 연령에 획일적인 정의를 내리지 않아서 국가별로 다름.
        - ILO기준을 따르는 우리나라는 구직활동 요건을 1999년 6월 이전까지는 1주기준, 그 이후부터는 4주기준으로 하고 이으며, 연령은 만 15세 이상인구를 기준으로 설정
        - 한편, ILO의 기준에 따르면 다음 세가지 요건을 모두 충족하면 실업자임.
            - 임금노동자 혹은 자영업자로서 자신의 일을 가지고 있지 않을 것
            - 기준 기간동안에 임금노동자 혹은 자영자로서 일이 제시되었다면 할 수 있었을 것
            - 지정된 기간 중에 고용이 되거나 자기사업을 시작하기 위해 구체적인 조치를 취했을 것
    
- **(지표작성)**
    - 실업자 여부를 결정하는데 자세한 내용은 `[그림 4-2]` 의 조사표를 참조
        - 먼저 취업자가 아닌 응답자 중에서 조사표 23번 문항의 "지난 4주내에 직장(일)을 구해보았습니까?"의 값이 1일 것
        - 이어서 조사표 24번 문항의 "지난주 직장(일)이 주어졌다면 일을 시작 할 수 있었습니까?"의 값이 1이면 실업자
    
- **(해석)**
    - 실업자와 비경제활동인구를 구별하는 기준과 실효성에 대한 다양한 연구가 존재
        - 경제활동과 비경제활동을 자주 넘나드는 '한계적 참가자'는 생산가능인구의 약 14%, 경제활동인구의 약 25%정도 되는 것으로 평가

### 2.4. 비경제활동인구

- **(정의)**
    - 만 15세 이상 인구 중 조사대상기간에 취업도 실업도 아닌 상태에 있는 자

- **(지표작성)**
    - 비경제활동인구 여부를 결정하는 자세한 과정과 내용은 `[그림 4-2]`의 조사표를 참조
        - 먼저 취업자가 아닌 응답자 중에서 조사표 23번 문항의 "지난 4주내에 직장(일)을 구해보았습니까?"의 값이 2일 것
        - 이어서 조사표 24번 문항의 "지난주 직장(일)이 주어졌다면 일을 시작 할 수 있었습니까?"의 값이 2이면 비경제활동인구
        
- **(해석)**
    - 생산가능인구 중에서 경제활동인구인 취업자와 실업자를 제외한 사람들로 주로 가사 또는 육아를 전담하는 주부, 학교에 다니는 학생, 일을 할 수 없는 연로자 및 심신장애인, 자발적으로 자선사업이나 종교단체에 관여하는 자 등이 주로 해당됨.
        - *구직단념자* 란 비경제활동인구중 취업을 희망하거나 취업가능성이 있지만 일정한 사유로 지난 4주간에 구직활동을 하지 않은 자 중 지난 1년내 구직경험이 있었던 자
            - 적당한 일거리가 없을 것 같아서(전공경력, 임금수준, 근로조건, 주변지역)
            - 지난 4주간 이전에 구직하여 보았지만 일거리를 찾을 수 없어서
            - 자격이 부족하여 (교육, 기술, 경험부족, 나이가 너무 어리거나 많다고 고용주가 생각할 것 같아서)
    - 추가적으로, **잠재취업가능자** 는 비경제활동인구 중에서 지난 4주간 구직활동을 하였으나, 조사대상주간에 취업이 가능하지 않은 자, **잠재구직자**는 비경제활동인구 중에서 지난 4주간 구직활동을 하지 않았지만, 조사대상주간에 취업을 희망하고 취업이 가능한 자

### 2.5. 고용률 및 실업률

- **(정의)**
    - 고용률은 만 15세 이상 인구 중 취업자가 차지하는 비율이며, 실업률은 실업자가 경제활동인구에서 차지하는 비율을 의미


$$고용률 = \frac{취업자수}{생산가능인구}\times 100$$
$$실업률 = \frac{실업자수}{경제활동인구}\times 100 = \frac{실업자수}{취업자수+실업자수}\times 100$$

- **(해석)**
    - 국내 노동시장의 일부 특수성을 감안하여 실업률과 고용률 지표를 함께 검토하는 것이 중요
        - 고령층 집단의 고용상황을 정확하게 판단하기 위해서 만 15~64세, 만 65세이상의 두 집단으로 구분하여 살펴보는 것이 바람직
        - 한편, 실업률을 해석하는 과정에서 높은 대학진학률과 장기 취업준비자가 많은 국내노동시장의 현실을 고려할 필요

### 2.6. 고용보조지표
 
- **(정의)**
    - 고용보조지표(Supplementary Indicators to Unemployment)는 일하고싶은 욕구가 충족되지 못한 노동력을 의미
        - 실업률이 노동시장의 고용상황을 정확하게 반영하지 못하는 한계를 보완하기 위한 보조지표
- **(지표작성)**
    - 국제노동기구(ILO)는 실업자 이외에도 취업자중에서 (1) 시간관련추가취업가능자와 비경제활동인구 중에서 잠재경제활동인구인 (2) 잠재취업가능자 및 (3) 잠재구직자 등을 이용하여 신규 고용보조지표를 설정
        - 고용보조지표는 노동공급과 수요가 일치하지 않아서 일하고 싶은 욕구가 있으나 일을 하고 있지 못하는 노동력을 나타내는 지표
        - 시간관련추가취업가능자란 조사대상기간에 실제 취업시간이 36시간 미만인 사람중에서 추가취업을 원하고 추가취업가능성이 있는 사람
        - 잠재경제활동인구란 비경제활동인구 중에서 취업에 관심을 표현한 사람으로 잠재취업가능자와 잠재구직자로 구성됨.
            - 잠재취업가능자란 비경제활동인구중에서 지난 4주간 구직활동을 하였으나, 조사대상기간에 취업가능성이 없었던 사람

### 2.7. 종사상 지위

- **(정의)**
    - 취업자가 실제로 일하고 있는 신분 또는 지위상태를 말하며 상용근로자, 임시근로자, 일용근로자, 고용원이 있는 자영업자(고용주), 고용원이 없는 자영업자(자영자), 무급가족종사자 등으로 구분
        - 임금근로자란 자신이 근로에 대해 임금, 봉급, 일당 등 어떠한 형태로든 일한 대가를 지급받는 근로자
        - 상용근로자란 고용계약설정자는 고용계약기간이 1년이상인 경우, 고용계약미설정자는 소정의 채용절차에 의해 입사하여 인사관리 규정을 적용받는 사람
        - 임시근로자란 고용계약설정자는 고용계약기간이 1개월 이상 1년 미만인 경우 고용계약미설정자는 일정한 사업(완료 1년미만)의 필요에 의해 고용된 경우인 근로자
        - 일용직근로자란 고용계약기간이 1개월 미만인 자 또는 매일 고용되어 근로의 대가로 일급 또는 일당제 급여를 받고 일하는 자 등
            - 상용근로자와 임시근로자는 경제사회노동위원회가 정의하는 정규직 및 비정규직과는 근본적으로 다른 개념 
            - 경제활동인구조사에서는 임시근로자와 일용근로자를 구별하여 조사하지만, 사업체노동력조사에서는 두 항목을 통합하여 조사
       - 고용원이 있는 자영업자란 한 사람 이상의 유급 고용원을 두고 사업을 경영하는 사람
       - 고용원이 없는 자영업자랑 자기 혼자 또는 무급가족종사자와 함께 자기 책임하에 독립적인 형태로 전문적인 업을 수행하거나 사업체를 운영하는 사람
       - 무급가족종사자랑 동일가구내 가족이 경영하는 사업체, 농장에서 무보수로 조사대상주간에 18시간 이상 일한 사람 
       
- **(경제활동인구 분해)**
    - 전체 내국인 장기체류 외국인  = 생산가능인구(15세이상인구) + 15세미만 인구
    - 생산가능인구(15세이상인구) = 경제활동인구 + 비경제활동인구
    - 경제활동인구 = 취업자 + 실업자
    - 취업자 = 임금근로자 + 비임금 근로자
        - 임금근로자 = 상용근로자 + 임시근로자 + 일용근로자
        - 비임금근로자 = 자영업자 + 무급가족종사자
            - 자영업자 = 고용주 + 자영자 
       
- **(해석)**
    - 종사상지위 중 자영업자에 대한 정의에 주의할 필요
        - 공식적으로 자영업자는 관리감독 여부를 기준으로 판단하는데, 업무를 자기책임 하에 독립적인 형태로 하면 자영업자로, 다른 사람의 책임하에 족속적인 형태로 지시를 받으면 임금근로자로 정의
        - 그러나, 실무적으로는 업무의 관리감독 여부와 상관없이 임금을 받으면 임금근로자, 임금을 받지 않으면 자영업자로 판단

        - 한국표준산업분류는 통계법에 의거하여 산업통계자료의 정확성과 국가간 비교성을 확보하기 위하여 모든 통계작성기관이 이를 의무적으로 사용

            - 현재 제10차 한국표준산업분류는 산업구조의 변화를 반영하기 위하여 UN 국제표준산업분류를 기초로 작성되었으며 2017년 개정됨

### 2.9. 표준직업분류

- **(정의)**
    - 수입(경제활동)을 위해 개인이 하고 있는 일을 그 수행되는 일의 형태에 따라 분류
        - 한국표준직업분류(KSCO: Korean Standard Classification of Occupations)는 직업관련 통계자료의 정확성 및 비교성을 확보하기 위하여 작성된 것으로 ILO의 국제표준직업분류(ISCO: International Standard Classification of Occupations)에 기초하여 제정
        - 현행 직업분류는 직업구조의 변화를 반영하기 휘아혀 2017년 7월 3일 제7차 개정을 고시(통계청 고시 2017-191호)하고 2018년 1월 1일부터 시행

- **(지표작성)**
   - 직업분류는 세분류르 기준으로 상위에는 대분류, 중분류, 소분류로 구성되어 있으며, 하위분류는 세세분류로 구성
       - 각 항목은 대분류 10, 중분류 52, 소분류 156, 세분류 450, 세세분류 1,231개로 계층적구조로 구성
       - 분류번호는 아라비아 숫자와 알파벳 A로 표시하며 대분류 1자리 중문류 2자리, 소분류3자리, 세분류4자리, 세세분류 5자리로 표시
       - 동일 분류에 포함된 끝 항목의 숫자 9는 '기타~(그 외~)'를 표시하며, 끝자리 0 은 해당분류수준에서 더이상 세분되지 않는 직업임.
       -  표준직업분류도 장기시계열 자료로 연결하여면 잦은 개정 때문에 원자료를 사용하는 추가적인 연결 작업이 필요하며, 자세한 내용은 통계청의 통계분류 홈페이지(https://kssc.kostat.go.kr:8443/ksscNew_web/kssc/main/main.do?gubun=1)를 참조


## 3. 임금

### 3.1. 사업체노동력조사

- **(정의)**
    - 사업체노동력조사는 현원, 빈일자리 및 입.이직에 관한 사항과 고용, 임금 및 근로시간에 관한 사항을 매월 조사하여 변동추이를 파악하고, 임금과 고용관련 정책의 기초자료를 제공하는데 그 목적이 있음.
    
- **(지표작성)**
    - 사업체노동력조사 근로실태부문의 자료는 다음과 같은 방법으로 작성되고 있음.
        - 임금총액(또는 총액임금)은 정액급여, 초과급여, 특별급여의 합으로 구성되며, 정액급여와 초과급여만 합하여 월급여(또는 월급여총액)라고 정의
            - 정액급여는 기본급, 통상적수당, 기타수장 등이며, 이때 통상적수당에는 직무수당, 기타수장에는 가족수당이 있음.
            - 초과급여는 연장근로수당, 야간근로수당 및 휴일근로수당 등이 해당
            - 특별급여는 상여금, 성과급, 임금인상소급분, 학자금(대출금 제외) 등으로 지급한 총액
            - 이에 따라 임금총액은 월급여액(정액급여+초과급여)+{전년도급여액(상여금+성과급)/12}로 계산
        - 소정근로시간이란 사업체의 취업규칙, 단체협약 등으로 정한 소정근로일에서 정규적인 업무개시 시각과 종료시각과의 사이에 실제로 근로한 시간으로 휴게시간은 제외
        - 초과근로시간이란 임금지급 여부와 관계없이 소정근로시간 이외의 시간(연장근로시간, 휴일근로시간)에 실제로 근로한 시간 (초과수당 지급을 위하여 산정한 시간이 아님)의 총계
        
- **(해석)**
    - 사업체노동력조사는 고용부문과 근로실태부문으로 구분되며, 근로실태부문을 주로 참고
        - 2011년부터 사업체고용동향조사(고용부문)와 사업체임금근로시간조사(근로실태부문)를 통합하여 사업체노동력조사로 명칭을 변경하였으나, 자료는 여전히 별도로 조사하여 공표되고 있음.
            -  고용부문은 종사자 1인 이상 사업체를 대상으로 조사하고 근로실태부문은 상용근로자 5인이상 사업체를 대상으로 조사하는 등 조사대상 자체가 다른 사실에 주의할 필요
        - 전체근로자는 상용, 임시, 일용근로자로 분류되며, 산업별 분류는 경제활동인구조사와 동일하게 10차표준산업분류를 사용
        - 사업체노동력조사의 근로실태부문(임금)은 5인이상규모의 사업장에 종사하는 전체근로자에 한정하고 있기 때문에, 1인이상 규모의 사업장에 종사하는 전체 근로자의 임금을 분석하기 위해서는 고용형태별근로실태조사, 경제활동인구조사 근로형태별부가조사 등을 사용하야 함.
    - 사업체노동력조사는 월평균을 월별 근로자수로 가중평균하여 분기와 연간 값을 계산
        - 다만, 근로자수를 고용노동부에서 공개하지 않기 떄문에 사실상 직접 계산이 불가능하고 고용노동부에서 제공하는 자료를 사용
        - 조사대상이 다르기 떄문에 가중치를 경제활동인구조사에서 근로자수로 대체할 수도 없음.

        - 두 조사는 고용형태별근로실태조사라는 명칭으로 통합되었으나, 임금구조기본통계조사와 고용형태별근로실태보사(구 사업체근로실태조사)는 독립적으로 조사 및 공표되는 자료임
            - 고용형태별근로실태조사는 임금근로자 1인이상 민간부문의 전산업을 대상으로 표본을 추출하기 떄문에 포괄적인 장점이 있으나, 2006년 이후부터만 이용가능하다는 단점이 있음.
            - 임금구조기본통계조사는 상용근로자 5인이상 민간붐ㄴ의 전산업을 대상으로 표본을 추출하기 때문에 임시일용근로자, 비임금근로자, 외국인근로자 등은 조사대상에서 제외되는 한계가 있으나, 원자료는 최소한 2000년이나 그 이전부터 가능하다는 장점이 있음.
