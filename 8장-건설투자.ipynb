# 8장 건설투자

## 1. 개요

- 국민계정상 건설투자는 국가 전체 건설부문의 투자흐름을 파악하는데 유용한 지표이나, 분기별로 공표되는 제약 때문에 건설부문에 대한 판단을 위해서 월별지표도 함께 고려
    - 실질 및 명목 원자료외에 계절조정 자료도 활용하여 건설투자 여건을 평가
- 국내 건설 관련 실적 통계뿐만 아니라, 거래와 심리지표등을 함께 이용하여 전반적인 건설경이 국면을 파악
    - 최근 동향을 판단하고 향후 건설경이를 전망하기 위해 건설투자 관련 동향 및 선행지표를 모두 활용

## 2. 국민계정 건설투자 및 건설업

### 2.1. 건설투자

**(정의)**
- 건설투자는 당해기간에 소진되지 않고 여러 회계기간에 걸쳐 생산에 반복적, 지속적으로 사용되는 유형자산의 순취득액 및 가치증가액을 포함하는 개념
    - 국민계정상 건설투자는 지출국민소득 측면에서 총고정자본형성의 하위분류이며, 총자본형성은 생산주체에 의한 고정자산의 취득을 의미
    - 사회간접자본이나 주택 등의 건설에 투자하는 것으로, 대규모의 자금이 투입되면서 경제성장 및 고용창출 효과가 매우 큰 편임.
        - 건설투자는 산출액, 물량기준으로 세금과 원자재 등을 포함 

**(지표작성)**
- 건설투자는 기본적으로 생산측면의 건설업 산출액 결과를 이용하며, 기준년가격(실질)과 당해년가격(명목)으로 나누어 추계
    - 기준년가격은 건설 총산출 및 영림(과수 등 육성자산) 총산출에 부대비용을 합산하여 구매자가격기준 총산출액을 구한 후 건물수선 및 건설재고 증감을 가감하여 추계
        - 기존 자산의 소유권이전에 따른 비용, 신규 건설자산의 분양시 발생되는 분양 수수료 수입 등을 포함하여 부대비용을 산출
        - 건설재고에는 소유권이 이전되지 않은 미분양주택 등이 포함되며, 건설재고잔액이 늘어나면 건설투자가 줄어들고, 반대의 경우애는 건설투자가 늘어남.
    - 당해년 가격은 통계청 건설경기 동향조사의 건설기성액을 주로 이용하여 추계
        - 명목 부대비용의 경우, 건설업 디플레이터 및 한국주택공사의 지가지수 등을 실질금액에 적용하여 환가
    - 계절조정은 미 상무부의 X12-AMRIMA를 우리나라 실정(설, 추석과 같은 사회적 관습 등)에 밎도록 조정한 방법을 이용하고 있으며, 하위부문의 계절조정 자료를 합산하여 상위부문의 계절조정 자료를 집계
        - 매년 연간 잠정 국민계정 발표시 직전 5개년 시계열을 수정

**(해석)**
- 생산자에 의한 고정자산 취득을 집계하므로 국가 전체의 건설투자 흐름을 파악하는데 유용
    - 국내총생산(GDP)과 직접적으로 비교$\cdot$분석이 가능하기 때문에 경제적인 측면에서 건설정책 방향 설정이 용이
    - 건설투자가 해당기간에 건설된 대상물을 소유하기 위해 소유자가 지출한 총 비용이라면, 건설기성액은 건축물 생성과정에서 생산자에게 추입된 총금액임.
        - 건설투자는 특정공식에 의해서 추정되는 수치로서 자산의 특성에 따라 일부 제외되거나 추가되는 항목이 있음.
    - 한편, 건설투자 디플레이터를 이용하여 건설투자 흐름을 함께 비교
        - 건설투자 디플레이터는 명목값으로 나누어 사후적으로 얻어지는 값으로, 건설투자에 영향을 주는 모든 물가요인을 포괄하는 종합적인 물가지수이며, 건설투자라는 상품의 가격수준을 나타낸다고 할 수 있음.

    - 한국은행은 보도자료를 통해 공표하고, 경제통계시스템에서도 관련자료를 모두 제공

### 2.2. 건설업

**(정의)**
- 국민계정상 건설업은 생산국민소득 측면에서의 산업활동 개념으로 정의 되며, 용도별로는 건물건설과 토목건설로 구분
    - 건설업자가 계약 또는 자가계정에 의하여 건설용지에 각종 건물 및 구출물을 신축, 증축, 해제하거나, 지반조성을 위한 공사를 수행하는 산업활동
        - 건설업은 세금과 원자재등이 제외된 부가가치 기준으로 건설투자보다 작은 개념
        - 다른 산업의 건설수요에 따라 생산활동이 이루어자는 파생산업으로 생산 결과는 유형의 실물이지만, 건설행위 자체는 실물을 생산하는 서비스 제공 행위임.
    - 건물건설(건물수선 포함)은 주거용과 비주거용 건물로 세분되어 단독주택, 공장 등을 포함하며, 토목건설은 도로, 전력 등 사회간접자본시설을 포괄함.

**(지표작성)**
- 연간 추계시에는 기본적으로 이중환가법(double deflation method)방식을 따르나, 분기에는 기초자료의 부족으로 전년도 부가가치율을 이용하여 추계
    - 기초자료인 건설기성액과 공사면적, 정부 건설투자 지출액, 민간 토목건설 지출액으로 명목 총산출을 추계한 후 공사종류별 디플레이터로 나누어 실질 총산출을 계산하고, 여기에 부가가치율을 이용하여 실질 부가가치를 산출
    - 당해년가격(명목)의 경우 건물건설과 건축보수, 토목건설의 총산출 추계에 차이가 있으며, 중간투입 및 부가가치 추계는 세 부문이 동일
        - 건물건설 총산출은 용도별 공사면적(실질)에 공종별 디플레이터를 적용하여 구한 공사금액(명목) 증감률과 건설경기동향조사의 건설기성액(명목)증감률을 이용하여 추계
        - 건축보수 총산출의 경우 통계청 가계동향조사의 주택유지 및 수선 지출액, 세대수, 건물수선재 내수출하량 등을 이용하여 추계
        - 토목건설은 통계청 건설기성액, 정부의 공사종류별 예산비율 및 잠정 예산집행액, 주요발주기업으로부터 조사된 총공사금액 등을 이용하여 추계
        - 중간투입은 실질 중간투입에 중간투입 디플레이터를 적용하여 추계하며, 부가가치는 총산출액에서 중간투입을 차감하여 산출
    - 기준년가격(실질)의 경우 건물건설, 선축보수, 토목건설 모두 동일한 방법으로 추계
        - 실질 총산출은 명목 총산출액을 총산출 디플레이터로 실질화하여 산출하며, 총산출 디플레이터는 중간투입 디플레이터와 부가가치 디플레이터를 가중평균하여 구함
        - 중간투입은 총산출액에 중간투입률을 곱하여 구하며, 부가가치는 총산출액에서 중간투입을 차감하여 산출
        - 다만, 토목건설의 경우 실질화시 각 공종에 해당되는 총산출 및 중산투입 가격지수를 사용하며, 부가가치율은 공종별 부가가치율을 적용
    - 지출측면의 건설투자와 동일하게 계절조정과 원계열 자료를 각각 발표

**(해석)**
- 전국 전체 건설업 활동의 흐름을 파악하는데 유용
    - 용도별로 주거용 건물건설, 비주거용 건물건설, 토목건설, 건물수선으로 구분된 자료가 제공되어 미시적인 분석이 가능

## 3. 건설기성과 건축착공

### 3.1. 건설기성액

**(정의)**
- 민간건설업체의 실제 시공실적을 평가한 것으로 일정기간 완료되니 공사에 대한 대가를 지급한 금액
    - 공사대금의 청구나 수취여부와는 무관하며, 공사비중 지가를 제외하고 발주자 공급원자재비와  부가가치액을 포함한 금액
        - 건설기성 실적은 부가가치액이 포함된 도급기성을 기준으로 하며 도급기성이 계산되지 않은 직영공사의 경우 실행기성을 조사함
            - 도급기성: 계약공사의 경우 공사실적에 대한 평가 금액으로 공사원가에 부가가치액을 합한 금액
            - 실행기성: 재료비, 노무비, 현장경비 등을 포함한 실제 시공에 소요된 공시원가 
    - 경상금액의 경우 발주자별로는 공공과 민간, 민자, 국내 외국기관으로, 공사종류별로는 건축과 토목으로 분류


**(지표작성)**
- 국내 공사에 한정하며, 전전년 *건설업조사* 결과 기준으로 누계기성액 상위 50%에 속하는 일반(종합)건설업체들을 대상으로 조사
    - 일반(종합)건설업체를 모집단으로 하는 표본조사이며, 표본집단의 크기는 2015년 기준 381개의 사업체로 구성
    - 경상금액의 최근 2개월, 불변금액(계절조정 포함)의 최근 3개월 수치는 잠정치임.
        - 불변금액의 경우 기준년도(현재 2015년)가 매 5년 주기로 변경
    - 계절조정 자료의 경우 매년(2월) 계절조정계열 확장으로 연간 보정시 과거계열이 변경됨.

### 4.1. 건설수주액

**(정의)**
- 일반적으로 건설업체가 발주자와 체결한 공사계약금액이며, 주로 건설부문의 선행지표로 이용됨.
    - 입$\cdot$낙찰 단계 이후 계약 단계에서 계약체결로 건설업자가 건설서비스를 제공함으로서 받는 대가의 총액
    - 최종계약일 기준으로 조사하며 공사계약금액 뿐만 아니라 매출부가가치세액(10%)과 발주자 공급원자재비를 포함한 금액이며, 토지보상비는 제외됨. 

**(지표작성)**
- 국내 공사에 한정하며, 전전년 *건설업조사* 경과 기준으로 누계 기성액 상위 54%에 속하는 일반(종합)건설업체들을 대상으로 조사
    - 일반(종합)건설업체를 모집단으로 하는 표본조사이며, 2015년 기준 742개의 사업체로 구성
    - 발주자별 건설수주액(경상), 공종별 건설수주액(경상)은 조사된 수주건별 100억이상 수주에 한에서도 별조로 집계
        - 발주자/공종별 건설수주액(경상)과 공종별 건설수주액(계절조정) 자료만 활용
    건설수주 공종분류가 2010년 변경됨에 따라 변경된 공종분류를 적용하여 과거 계열을 재집계

**(해석)**
- 건설수주는 건설업자의 매출 및 건설기성을 사전에 예상할 수 있어 건설경기를 판단하기 위해 사용되는 가장 일반적인 선행지표임.
    - 그러나, 집계시점에 따라 건축변경 및 취소 등의 이유로 건설수주액이 모두 건설기성으로 연결되는 것은 아니며, 대규모 건설공사의 수주가 불규칙하게 발생함애 따라 상대적으로 다른 지표들에 비해 월별 변동 폭이 큰 편임.
    - 한폄, 건설수주는 건축허가면적과 비교했을 때 변화 방향이 서로 다르게 나타나는 경우도 있음
        - 이는 건설수주는 건축과 토목 모든 부문의 건설공사계약금액을 조사하는 반면, 건축허가면적은 건축부문만을 대상으로 허가면적을 집계하는 것에 기인

### 4.2. 건축허가면적

**(정의)**
- 건축행위를 직접 규제하고 있는 법령들에 적합한지 여부에 대한 확인을 거친 건축물의 건축허가면적
    - 건축과 토목으로 이루어진 건설부문 중 민간부문이 큰 비중을 차지

**(지표작성)**
- 건축착공면적과 동일

**(해석)**
- 건축허가 현황을 파악하여 건설경기 및 주택수급의 동향 관련 분석은 물론 건설경기 전망을 위한 기초통계 자료로 활용
    - 주거용 및 상업용, 공업용 등 용도별로 세분화된 통계자료를 제공
    - 일반적으로 매년 4/4분기는 전분기에 비해 큰폭의 증가를 보임.
        -착공시기가 가장 유리한 1/4분기 또는 2/4분기를 고려하여, 사업 승인 신청시점이 아파트의 경우 전년도 하반기에 집중되는데 기인
    - 집계시점에 따라 건축변경 및 취소 등의 이유로 수치가 변동될 수 있어, 시점별 절대수준의 비교에 주의할 필요
    - 건축허가면적 중 주거용의 경우 주택건설실적(인허가 실적)통계와 유사성이 있으나, 공표단위가 달라 절대적인 증감률 비교에는 한계가 있음.
        - 건축허가면적은 면적 단위인 반면, 주택건설인허가 실적은 호수 단위임

**(출처 및 공표일시)**
- 건축착공면적과 동일

**(조사주기 및 자료의 형태)**
- 매월 조사하며, 2002년 1월부터 월, 1990년부터 분기 및 연간 자료로 이용 가능 
    - 조사기간과 단위는 건축착공면적과 동일

### 4.3. 주택건설인허가실적

**(정의)**
- 주택건설사업승인과 주택건축허가의 실적을 합산한 것으로 주택경기를 판단하는 과정에서 선행지표로 활용
    - 주택건설에서 사업승인 대상은 주택법 제15조에 의한 30호(세대) 이상 규모의 단독 또는 공동주택이며, 30세대 미만 주택규모는 건축법 제8조에 의한 건축허가 대상임.
        - 주택과 주택외의 시설물이 하나의 건축물로 된 주상복합건축물은 300세대 이상의 경우 주택사업승인 대상이고, 300세대 미만일 경우, 주택이 차지하는 ㅁㄴ적이 총 연면적의 90% 미만인 경우에 한해 건축허가대상임.

**(지표작성)**
- 주택건설착공실적과 동일

**(해석)**
- 주택경기 관련 선행지표로 전반적인 부동산시장의 흐름을 예측하는데 유용
    - 규모별 통계와 함께 주택유형별 및 지역별 통계와 같은 세분화된 정보도 함께 제공

**(출처 및 공표일시)**

### 6.1. 한국은행 기업경기실사지수(BSI)

**(정의)**
- 기업들의 경영활동 실적과 전망을 파악하여 지수화한 통계
    - 기업의 생산, 매출, 투자 등에 중대한 영향을 미치는 기업가의 경기에 대한 판단 및 전망 등을 설문조사하여 작성되는 지수
        - 100보다 클 때는 경기판단에 대한 긍정적인 시각의 기업 수가 부정적인 시각의 기업 수보다 많다는 의미

**(지표작성)**
- 2013년 기준 국세청 법인세 신고 기업 중 업종별 매출액 누적액 기준 상위 95% 업체를 대상으로 우편 및 전화조사
    - 2016년 7월 기준 전국 3,313개 법인기업으로, 금융, 보험 및 외국기업 국내지사 등은 제외
        - 제조업은 23개 중분류 업종, 비제조업은 13개 대분류 업종
    - 각 업체의 응답을 아래의 공식에 따라 지수화
        - $$업종별BSI = \frac{긍정적인응답업체수 - 부정적인응답업체수}{전체응답업체수}\times 100 + 100$$
        - $$산업(제조업,비제조업)별 BSI = \sum_{i=1}^{n}w_i\cdot BSI_i$$
            - $w_i$: 각업종별 GDP 비중, $BSI_i$: 각업종별 BSI
        - 전산업, 제조업 및 비제조업, 서비스업 지수 편제 시 ㅣ업종별 GDP 비중을 가중치로 사용

**(해석)**
- 기업경기실사지수는 실물지표와 높은 상관관계를 갖고 있어 경기 동향을 판단하고 전망하는데 매우 유용한 지표
    - 업황, 제품제고, 인력사정, 매출규모, 생산규모, 신규수주규모, 가동률, 자금사정 등에 대한 당월 실적 및 익월 전망 자료를 이용
        - 보통 업황실적(전망), 매출실적(전망), 채산성실적(전망), 자금사정실적(전망)
, 인력사정실적(전망) 자료를 이용

**(출처 및 공표일시)**
- 한국은행에서 매월 보도자료를 발표하며, 경제통계시스템을 통해서도 관련 자료를 제공
    - 매월 말 한국은행 경제통계시스템에서 다운로드
        - 한국은행 BSI 경로: https://ecos.bok.or.kr/ ---- 통계검색 ---- 6. 심리지수 ---- 6.1. 기업경기실사지수 ---- 6.1.2. 업종별 기업경기실사지수, 2009.8~ ---- 6.1.2.1. (실적), 6.1.2.2.(전망) ----

### 6.2. 전국경제인연합회 기업경기실사지수(BSI)

**(정의)**
- 각 기업의 전월 실적, 기업가의 익월 경기 동향에 대한 의견이나 계획 등을 조사하여 지수화한 것
    - 조사방식이 설문조사에 의존하므로 실물경기에 민감하고 여타의 자료로는 포착할 수 없는 기업 활동에 대한 평가와 전망을 수량화할 수 있다는 것이 장점
    - 한국은행에서 발표하는 기업경기실사지수와 개념은 동일하나, 금융업 등 경기대응성이 낮고 조사가 곤란한 일부 업종은 제외

**(지표작성)**
- 업종별 매출액순 600대 기업을 대상으로 응답기업 담당자의 자기기술과 조사원의 질의기술을 병행하여 조사하며, 계산방법은 한국은행의 기업경기실사지수와 동일
    - 조사대상 업체는 한국신용평가의 전년도 매출액 기준 2,000대 기업을 표본으로 선정
    - 설문은 8개 항목으로 구성되며, 설문방식도 판단조사만을 실시
    - 기업규모를 고려하지 않을 경우 지수가 왜곡될 가능성이 존재하기 때문에, 기업의 매출액 비중을 기준으로 재산정하여 가중BSI를 발표
    - 추세변동, 계절변동, 순환변동, 불규칙변동에 대한 조정과정을 거쳐 계절조정지수를 산출

**(해석)**
- 기업의 경영활동에 대한 기업가의 경기판단과 전망을 실물지표와 비교할 수 있는 지표
    - 한국은행 경기실사지수가 조사자료에 근거한 정량적 분석이라면, 전경련 경기실사지수는 평가의 원인과 배경도 제공
    - 산업별 분류와 함께 종합경기, 내수, 투자, 수출, 자금사정, 재고, 고용, 채산성의 부문별 고사가 병행됨.
    - 한국은행 기업경기실사지수와 전경련 기업경기실사지수의 추이는 비슷한 모습이나 대체로 전경련의 통계가 한국은행 통계보다 더 높은 수준에서 형성

**(출처 및 공표일시)**
- 전경련 홈페이지에서 보도자료의 형태로 매월 하순경 공표
    - 조사 결과에 대한 원인 및 배경은 전경련 홈페이지에서 볼 수 있으며, 부문별로 자세한 수치는 아래 경로를 통해 확인 가능
        - 전경련 BSI 경로: https://www.fki.or.kr/ ---- 전경련소식 ---- 보도자료$\cdot$발표문 ----[BSI] 검색

### 6.3. 한국건설산업연구원 건설기업 경기실사지수(CBSI)

**(정의)**
- 국내 종합 건설업체의 기획담당자를 대상으로 국내 건설기업의 체감경기를 조사해 지수화한 건설기업 경기실사지수
    - 한국은행과 전경련이 업종별로 분류된 건설업 경기실사지수를 조사하는 반면, 한국건설산업연구원은 건설기업에만 한정하여 작성

**(지표작성)**
- 응답업체가 매번 동일한 패널 자료로, 온라인 설문조사 방식을 통해 집계
    - 매출액을 기준으로 대기업, 중견기업, 중소기업으로 구분하여 조사하며, 조사참여업체는 약 400여개임.
        - 2000년 처음 조사 시 설계된 기업순위 및 각 업체 규모별 시장점유율 등을 토대로 대형, 중견 ,중소건설업체로 구분
    - 지수 산출 방식은 한국은행과 동일하며, 각 업체 규모별로 산출된 지수에 각각의 시장점유율을 가중치로 곱한 후 이 값을 합산.

**(해석)**
- 경제주체의 체감경기와 심리를 직접 조사하여 현재의 경기 상황을 파악하고 미래경기를 예측하는데 사용되는 심리지표
    - 지수 100을 기준으로 100이상이면 경기를 긍정적으로 체감하고 있는 응답업체가 더 많고, 100 미만이면 경기를 부정적으로 체감하고 있는 응답업체가 더 많은 상황
    - CBSI외에 한국건설산업연구원에서 발간하는 월간건설경기동향과 건설동향브리핑(주간) 자료도 함께 활용하면 건설경기 동향을 파악하는데 유용
    - 다만, 건산연 CBSI지수는 한국은행과 전경련의 자료에 비해 발표가 다소 늦는 경향이 있음.

**(출처 및 공표일시)**
- 건설기업 경기실사지수는 매달 첫째 주에 공표
    - 발표 후 한국건설산업연구원에서 PDF파일 형태로 이용 가능
        - 한국건설산업연구원 CBSI 경로: http://www.cerik.re.kr/ ---- 연구원발간물 ---- 시장전망
