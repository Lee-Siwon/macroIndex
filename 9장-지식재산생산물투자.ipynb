# 9장 지식재산생산물 투자

## 1. 개요

- 지식재산생산물이 장기간 우리 경제의 생산성 향상에 크게 기여하는 등 최근 중요성이 더욱 증대됨에 따라 지식재산생산물투자와 연구개발관련 서비스업생산지수 등 종합적으로 분석
    - 국민계정상의 지식재산생산물투자를 연구개발과 기타지식재산생산물투자로 분류하여 살펴보고, 통계청의 산업별 서비스업생산지수 중 연구개발서비스업과 출판, 영산, 방송통신 및 정보서비스업 부문을 점검

## 2. 지식재산생산물투자

**(정의)**
- 지식재산생산물투자는 1년 이상 장기간 동안 생산에 사용될 것으로 예상되는 연구개발(R&D)과 오락$\cdot$문학작품 및 예술품 원본, 컴퓨터 소프트웨어 및 데이터 베이스, 광물탐사 등의 기타지식재산생산물투자로 분류 
    - 종전에 비용으로 처리되었던 지식재산생산물투자가 2014년 국민계정 개편 시 새로운 국제기준(2008SNA)이 적용되어 자산으로 변경됨

**(지표작성)**
- 연구개발은 한국과학기술평가원이 조사한 연구개발활동조사보고서의 경상비와 자본적 지출 등 연구개발비 자료에 기초하여 추계
    - 이때 국민계정 통계와 조사보고서 간에는 개념과 포괄범위 등에 차이가 있기 떄문에 조사보고서를 국민계정체계 기준에 맞게 조정하는 과정을 거쳐 생산액을 산출
- 기타지식재산생산물의 경우 국내생산액은 통계청, 각종 협회 등의 기초 통계 자료를 이용하여 추계
    - 국내 생산액과 수입액에 각각 산업연관표 상의 중간소비 비중에 의거 최종 수요 금액을 산정한 후 상품흐름법을 적용하여 생산자가격 기준 투자 금액을 계산하고 운송 및 거래 마진을 합산하여 구매자 가격 기준 투자금액을 산출

**(해석)**
- 지식재산생산물투자는 생산성 향상에 크게 기여한다는 점에서 우리 경제의 성장 여부를 판단할 수 있는 매우 중요한 실물지표 
    - 연구개발 등 지식재산생산물이 자산으로 처리되면서 영업잉여(부가가치)와 투자, 그리고 GDP가 증가

**(출처 및 공표일시)**
- 지식재산생산물투자는 한국은행의 국민계정 또는 분기 국내총생산 보도자료에 포함되며, 경제통계시스템에서 관련 자료를 모두 제공
    - 한국은행 경제통계시스템 홈페이지에서 공표일정을 제공하고 있음.
        - 지식재산생산물투자 경로: https://ecos.bok.or.kr/ ---- 통계검색 ---- 2. 국민계정 ---- 2.1. 국민소득통계(2015년 기준년) ---- 2.1.6. 총자본형성 ---- 2.1.6.1. 자본재별 총자본형성 ---- 지식재산생산물투자

**(조사주기 및 자료의 형태)**
- 1970년 이후부터 연간, 분기별 지표의 원계열/계절조정계열 및 명목/실질 자료를 발표

## 3. 연구개발서비스업 및 출판, 영상, 방송통신 및 정보서비스업

**(정의)**
- 국민계정상의 연구개발과 기타지식재산생산물투자와 연관된 월별 연구개발서비스업 생산지수와 출판, 영상, 방송통신 및 정보서비스업 생산지수를 활용

**(지표작성)**
- 자세한 내용은 *경기*의 서비스업생산지수를 참고

**(출처 및 공표일시)**
- 통계청에서 매월 말일이나 마지막 주 금요일에 전월의 잠정치를 발표하고, 2개월 후 확정치를 발표
    - 국가통계포털의 주제별 통계에서도 관련 자료를 제공
        - 연구개발서비스업 경로: https://kosis.kr/ ---- 국내통계/주제별통계 ---- 도소매$\cdot$서비스 ---- 서비스업동향조사 ---- 산업별 서비스업생산지수 ---- 연구개발업(2레벨)
        - 출판, 영상, 방송통신 및 정보서비스업 경로: https://kosis.kr/ ---- 국내통계/주제별통계 ---- 도소매$\cdot$서비스 ---- 서비스업동향조사 ---- 산업별 서비스업생산지수 ---- 출판, 영상, 방송통신 및 정보서비스업

**(조사주기 및 자료의 형태)**
- 2000년부터 연간, 분기, 월별 지표의 경상지수, 불변지수, 계절조정지수를 발표하며 13개의 업종별로 세분류
