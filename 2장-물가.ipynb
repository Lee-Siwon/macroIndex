# 2장 물가

## 1.개요

- 물가지수는 본래 상품과 서비스의 가격변동을 측정하기 위한 수단이지만, 경기판단의 지표로서의 역할을 수행하며, 또한, 주요 경제지표의 디플레이터로 활용
    - 우리나라의 대표적인 물가지표로는 소비자물가지수, 수출입물가지수, 생산자물가지수, 주택가격지수, 지가지수 등이 있음
    - 모든 물가지수는 월별자료이므로 매월 수치를 취합하여 분석하되, 분기별로 자료를 분석할 때에는 해당 월별지수의 평균 혹은 기말기준 값을 사용
    - 계절성 및 일시적 충격 등의 요인을 주의하고, 분석 목적에 따라 원지수, 전기대비 상승률, 전년동기대비 상승률을 선택하여 사용

## 2. 소비자물가

- **(정의)**
    - 소비자물가지수는 소비자가 일상생활을 영위하기 위해 구입하는 상품과 서비스의 가격변동을 측정하여 지수화한 것을 의미
        - 주요 품목의 실제 거래가격을 측정함으로써 최종 소비단계의 물가수준을 종합적으로 판단하는데 활용

- **(지표작성)**
    - 소비자가 구입하는 상품과 서비스 품목을 대상으로 가중치를 부여해 합산
        - 2015년을 기준으로 가계소비지출에서 차지하는 비중이 1/10,000 이상인 품목 481개를 대상으로 작성하며, 각각의 소비지출액 구성비에 따라 1,000분비로 산출한 가중치를 부여하여 합산한 수치임..
        
**[표 3-2] 소비자물가지수의 종류**

|지수명|내용|
|:---|:---|
| 지출목적별지수 | 소비지출 목적에 따라 분류 (12개 대분류) |    
| 품목성질별지수 | 조사대상 품목의 특성에 따라 상품(327개)와 서비스(154개)로 구분하여 작성 |    
| 농수산물및석유류제외지수 | 계절적인 요인이나 일시적인 충격에 의한 물가변동분을 제외하고 장기적인 추세를 파악하기 위해 곡물외의 농산물과 석유류 품목을 제외하여 작성(429개) |    
| 식료품및에너지제외지수 | 논산물과 석유류 외에도 축산물, 수산물 및 에너지 품목을 제외하여 작성(337개) |    
| 생활물가지수 | 체감물가를 설명하기 위해 구입빈도가 높고 지출비중이 높아 가격변동을 민감하게 인식하는 품목으로 작성(142개) |    
        
자료: 통계청 보도자료 *소비자물가동향조사*

- **(해석)**
    - 매월 상품과 서비스 가격의 변동률을 조사하여 물가상승에 따른 소비자부담, 구매력 등을 측정
        - 소비자물가지수는 물가변동이 도시가구의 소비행활에 미치는 영향을 나타내는 지표로 어느 특정 가구나 계층을 대상으로 측정하는 것이 아니라 전체 도시가구의 평균적인 영향을 나타냄
        - 활용목적에 따라 여러가지 분류방법이 있으나, 품목 성질별 지수와 농산물 및 석유류제외 지수(근원물가지수)를 주로 사용
            - 품목성질별지수는 상품 또는 서비스 유형별 물가변동을 파악하는데 용이
            - 농산물및석유류제외지수(근원물가지수)는 일시적 공급 충격의 영향을 많이 받는 곡물 외의 농산물과 석유류 품목을 제외한 것으로 소비자물가의 장기적인 추세를 잘 반영함.
        - 한편, 소비자물가의 변화 추이를 점검할 때, 개별 품목의 기여도도 함께 활용

## 3. 수출입물가

- **(정의)**
    - 수출입물가지수는 수출입 상품의 가격변동을 조사하여 국내물가에 미치는 영향을 측정하는데 이용되는 지표


- **(지표작성)**
    - 국내물가에 미치는 영향을 용이하게 파악할 수 있도록 수출입 계약가격을 원화로 환산하여 작성한 원화기준지수를 주지수로 편제하고, 계약통화기준지수와 달러화기준지수를 보조지수로 작성
        - 국내물가에 대해 선행성이 반영도리 수 있도록 수출입계약시점을 기준으로 하되 원칙적으로 월평균 가격을 조사
            - 수출물가는 FOB가격, 수입물가는 CIF가격을 기준으로 하지만, 일부 품목의 경우 거래관행을 적용
        - 연쇄로우(Lowe)산식을 사용하되 무역구조의 변화를 반영하기 위해 매년 조사개상의 품목 및 가중치를 변경
        - 조사대상은 개별품목의 수출입액이 모집단거래액의 1/2,000이상의 거래비중을 차지하고 동종 상품군의 가격변동을 대표하고 가급적 품질규격이 균일하게 유지하되 가격시계열 유지가 가능한 품목을 선정(2023년 1월 현재 수출 213개 품목 -농림수산품 3개, 공산품 210개-, 수입 237개 품목 -농림수산품 18개, 관산품 12개, 공산품 207개-)
        
- **(해석)**
    - 수출입물가지수는 국내 수출입 기업의 전반적인 대외거래 상황에 대한 유용한 정보를 제공하는 한편, 교역조건 측정과 수출입액 실질화에도 이용
        - 수출입물가지수가 원화, 계약통화, 달러화 기준 등으로 작성되고 있는 만큼 각 수출입물가지수의 차이점에 대해 명확히 이해할 필요
            - 계약통화기준 지수는 수출입 상품의 세계 수요 및 공급 여건을 반영하는 한편, 원화기준 지수는 추가로 원화환율의 변동을 반영함에 유의
        - 수출물가와 수입물가를 분석하는 과정에서 각각 기본분류지수와 품목별 분류지수를 주로 사용
    - 수입상품의 국제가격 변동이 국내물가에 미치는 효과를 예측하거나 수출입관련업체들의 수출채산성 변동 및 수입원가부담 등을 파악하는 한편 수출입물가지수의 상호비교를 통해 가격측면에서의 대외교역조건 등을 측정하는 데에 이용

## 4. 생산자물가

- **(정의)**
    - 생산자물가지수는 국내에서 생산하여 국내시장에 출하되는 모든 재화와 서비스의 가격변동을 측정하는 지표
        - 즉, 국내 생산자가 재화와 서비스를 시장에 처음 출하하는 가격으로 소매업자가 지불하는 가격을 측정
        - 기본분류지수와 특수분류지수 등을 제공하며 주로 기본분류지수를 사용

- **(지표작성)**
    - 매년 물가지수의 가중치를 변경하여 경제구조의 변화를 신속히 반영하는 연쇄가중방식을 채택
        - 국내출하액이 모집단금액의 일정수준(상품 1/10,000, 서비스 1/2,000)이상의 거래비중을 갖고 동종 제품군의 가격변동을 대표할 수 있으며 가격 시계열 유지가 가능한 품목을 선정하고 있음 (2023년 1월기준 891개 품목 -상품:787개, 서비스:104개-)
        - 가중치는 선정된 품목과 유사품목의 국내출하액이 모집단금액에서 차지하는 비중으로 물가지수내 조사대상품목의 중요도를 시사
        - 한편, 생산자물가지수는 상품의 경우 소비재뿐만 아니라 자본재와 생산과정에 투입되는 원재료와 중간재까지도 포함하며, 서비스의 경우에는 일부 개인 서비스외에 주로 기업 서비스의 물가 변화를 반영
        
- **(해석)**
    - 생산자물가지수는 포괄 범위가 넓어서 전반적인 상품과 서비스, 수급동향, 물가수준 등을 종합적으로 반영
        - 생산자물가가 상승하는 경우, 소비자물가의 상승으로 이어지거나 특히 경기불황일 때에는 생산자 혹은 소매업자의 이익이 감소하기도 함
        - 생산자물가 동향을 판단할 때는 이동평균 등을 이용하여 방향성을 확인하는 것이 중요
        - 소비자물가지수와 비교는 가능하지만, 생산자판매단계와 소매단계의 마진을 측정하는 데 이용될 수 없음.

## 5. 주택가격

- **(정의)**
    - 주택가격지수는 전국 주택의 매매 및 전세가격을 바탕으로 전국주택시장 동향을 파악하는 지수

- **(지표작성)**
    - 전국 시$\cdot$군$\cdot$구의 거래가능한 아파트, 단독주택, 연립주택(임대 제외)을 대상으로 선정된 표본에 대해서 전문조사원들이 현장 조사
       - 표본주택이 저래된 경우에는 실거래가격을 사용하고, 거래되지 않은 경우에는 유사표본의 실거래가격을 반영
       - 기준시접 대비 조사시점 가격비의 기하평균방식으로 최소공표단위 지수를 계산하며, 주택재고량을 가중치로한 제본스(Jevons) 지수 산식을 이용하여 광역단위 지수를 계산
       - 주택가격지수는 KB국민은행에서 작성하는 국가승인통계업무였으나, 2012년부터 한국감정원으로 이관된 이후 조사대상, 통계 작성법 등이 변경됨.

| 통계 작성방법 | 라스파이레스(Laspeyres) 지수방식 | 제본스(Jevons) 지수방식(국제기준) |

- **(해석)**
    - 우리나라 주택시장의 평균 가격변화를 파악할 수 있는 통계로, 정부를 비롯한 관련 기관들에서 주택경기를 판단하는데 기초자료로 유용하게 사용
        - 기준시점(2022년 1월)의 주택가격을 100으로 한 상대가격을 통해 주택가격의 변화를 표시
        - 주택가격지수는 표본의 가격 비율로 산정되는 통계자료이기 떄문에 이 지수를 이용하여 지역간의 직접적인 가격 차이를 비교할 수는 없음.
        - 주택유형 중 전체의 약 80%를 구성하는 아파트 매매가격변동은 자산시장분석의 가장 중요한 요소임.
        
- **(출처 및 공표일시)**
    - 월별 전국주택가격지수는 한국감정원이 매월 1일(공휴일인 경우 그 다음날) 발표
        - 주택가격지수 경로: https://www.reb.or.kr/r-one/main.do ----공개자료실 ---- 통계자료받기 ---- 주택가격지수
        
- **(조사주기 및 자료의 형태)**
    - 주택가격지수는 월 1회 조사되며, 월별 지표의 원계열과 계절조정계열 자료를 발표하고 있음.
        - 주택가격 상승률의 경우 지수를 이용하여 직접 계산한 수치와 한국감정원의 공표자료가 불일치할 때 공표된 자료를 사용하는 것이 바람직

## 6. 지가

- **(정의)**
    - 지가변동률은 표본지의 평가가격을 기초로 하여 산정된 지가지수의 기중시점과 비교시점의 비율을 의미 
        - 전국의 지가변동 상황을 조사하여 토지정책 수행과 감정평가의 기초 자료로 활용하는 정부의 공식 통계
        
**[표 3-4] 지역별 지가**

| 구분 | 지역 |
|:----|:----|
| 대도시 | 서울특별시, 부산,대구,인천,광주,대전,울산광역시, 세종특별자치시 | 
| 시군지역 | 대도시를 제외한 지역 |  
| 수도권 | 서울특별시, 인천광역시, 경기도 |  
| 지방권 | 수도권을 제외한 지역 |  

- **(지표작성)**
    - 전국 250개 시군구 및 3,136개 기초구역(읍면동)의 80,000개 필지 표본에 대해서 전문조사원들이 현장조사한 후 가격을 평가
        - 연관성이 높은 지역(법정동)을 묶어서 구획한 기초구역을 단위로 조사하고, 표본지의 가격에 가중치를 적용하는 라스파이레스 수정 산식을 통해 작성
        
- **(해석)**
    - 전국 시도 및 지역별, 용도지역별, 이용상황별 지가변동률 등 다양한 수치를 바탕으로 분석
        - 지가통계는 일반적으로 지가지수의 변동을 측정하기 위한 지표로서 개별토지의 가격수준과 변동을 측정하는 것은 아님
        - 개졀토지의 감정평가, 보상액 및 매수가격 산정, 개발부담금 산정, 토지거래허가구역 지정 및 투기지역 지정 등 부동산 정책등의 토지정책 수행을 위한 기초 자료로 활용
        - 해정구역 변경 및 개발 사업등에 의한 용도비역 변경사항이 충분히 반영되지 못하고 표본 재설계를 통해 후행적을 반영되는 한계가 존재
        
- **(출처 및 공표일시)**
    - 전국지가변동률조사는 국토교통부의 보도자료를 통해 배포되며, 한국감정원 등을 통해 매월 하순 공표됨.
        - 지가 경로: https://www.reb.or.kr/r-one/main.do ----공개자료실 ---- 통계자료받기 ---- 지가변동률
        
- **(조사주기 및 자료의 형태)**
    - 지가는 1975년 이후부터 연간, 월별로 원계열 자료가 제공되고 있음.

## 7. 기타
- **(정의 및 해석)**
    - 월별 물가동향을 분석하는 과정에서 정확성 제고를 위해 기대인플레이션, 물가인식, 원/달러 환율, GDP갭률, 실효환율, GDP디플레이터, 유가등을 참고 지표로 활용함
        - 기대인플레이션의 경우, 설문조사에 기초한 기대인플레이션과 채권시장수익률을 통한 기대인플레이션을 이용
            - 기대인플레이션(설문조사)은 한국은행 소비자동향조사 결과 향후 1년간 소비자 물가상승률을 예측한 수치로서 가계가 예상한 물가상승률
            - 기대인플레이션(채권시장)은 물가연동채권과 장기채권(10년물)의 수익률차로 시장참여자들이 예측한 물가상승률
        - 물가인식은 지난 1년간 소비자물가상승률에 대한 일반인의 인식을 조사한 결과로서 과거 소비자들의 물가에 대한 인식을 나타내는 지수
        - 원/달러 환율은 원화와 외국통화의 교환비율로서 외국통화와 비교한 원화의 가치를 나타내며, 수출입물가에 미치는 영향을 파악하는데 주로 이용
            - 실질실효환율은 자국통화와 주요 교역상대국의 명목환율을 무역비중으로 가중평균하여 산출한 지수로서 자국상품의 가격경쟁력을 나타내는 지표
        - GDP갭률은 실제 GDP와 잠재GDP간의 차이를 잠재GDP로 나눈값으로, 경기변동(경기과열 및 경기침체)을 설명하는데 활용
        - GDP디플레이터는 명목GDP를 실질GDP로 나눈 것으로 통상 국민경제의 물가수준을 나타내며 각 물가지수와 임금, 환율 등 각종 가격지수가 반영되어 있음.
        - 유가(Dubai)는 중동산 석유가격의 기준으로 국제석유가격의 변동이 국내물가에 미치는 영향을 파악하는데 활용
